```python
import pandas as pd
from datetime import datetime

data = {
    'CustomerID': ['C1', 'C2', 'C3', 'C1', 'C2', 'C3', 'C1', 'C2', 'C3'],
    'TransactionID': ['T1', 'T2', 'T3', 'T4', 'T5', 'T6', 'T7', 'T8', 'T9'],
    'TransactionDate': [datetime(2024, 1, 1), datetime(2024, 1, 2), datetime(2024, 1, 3), datetime(2024, 2, 1), datetime(2024, 2, 2), datetime(2024, 2, 3), datetime(2024, 3, 1), datetime(2024, 3, 2), datetime(2024, 3, 3)],
    'ProductID': ['P1', 'P2', 'P3', 'P1', 'P2', 'P3', 'P1', 'P2', 'P3'],
    'TransactionAmount': [100, 200, 300, 400, 500, 600, 700, 800, 900]
}

df = pd.DataFrame(data)

df
```

```python
df['MonthYear'] = df.TransactionDate.dt.strftime('%m-%Y')
```

```python
df.groupby(['CustomerID', 'MonthYear'])['TransactionAmount'].sum().reset_index()
```

```python
import pandas as pd

def calculate_monthly_spend(df):
    df['MonthYear'] = df.TransactionDate.dt.strftime('%m-%Y')
    return df.groupby(['CustomerID', 'MonthYear'])['TransactionAmount'].sum().reset_index()
```

```python
calculate_monthly_spend(df)
```
