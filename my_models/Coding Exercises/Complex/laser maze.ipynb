```python
#Laser Maze by 101Comoputing.net - www.101computing.net/laser-maze/
import random, time, os, copy
from IPython.display import clear_output
          
#Initialise maze, bombs, laser source and exit gate

def drawMaze():
    
    for i in range(0,12):
        line = ""
        for j in range(0,12):
            line = line + maze[i][j] 
        print(line) 
        
        
def placeForwDefl(row,col):
    maze[row+1][col+1] = "///"   
     
def placeBackDefl(row,col):
    maze[row+1][col+1] = "\\\\\\"   
    
def shoot():
    
    global maze
    
    maze = [[""," ","  1","  2","  3","  4","  5","  6","  7","  8","  9"," "],
            [" "," ","---","---","---","---","---","---","---","---","---"," "],
            ["1","|","   ","   ","   ","   ","   ","   ","   ","   ","   ","|"],
            ["2","|","   ","   ","   ","   ","   ","   ","   ","   ","   ","|"],
            ["3","|","   ","   ","   ","   ","   ","   ","   ","   ","   ","|"],
            ["4","|","   ","   ","   ","   ","   ","   ","   ","   ","   ","|"],
            ["5","|","   ","   ","   ","   ","   ","   ","   ","   ","   ","|"],
            ["6","|","   ","   ","   ","   ","   ","   ","   ","   ","   ","|"],
            ["7","|","   ","   ","   ","   ","   ","   ","   ","   ","   ","|"],
            ["8","|","   ","   ","   ","   ","   ","   ","   ","   ","   ","|"],
            ["9","|","   ","   ","   ","   ","   ","   ","   ","   ","   ","|"],
            [" "," ","---","---","---","---","---","---","---","---","---"," "]]
    
    source = (random.randint(1,9)+1,1)
    row_source, col_source = source[0],source[1]
    maze[row_source][col_source] = ">"

    exit = (random.randint(1,9)+1,11)
    row_exit, col_exit = exit[0], exit[1]
    maze[row_exit][col_exit] = "="    
    
    drawMaze()
    
    direction = "right"
    started = False
    state="waiting_user_input"
      
    while state == "waiting_user_input":
        
        user_input = input('Select \nM: add mines \nF: add forward deflector \nB: add backward deflector \nS: start \nR: restart\n')
        
        if user_input=='M':
            state = 'adding_mines'
            while state == 'adding_mines': 
                try:
                    numb_mines = int(input('Insert the number of mines to add'))
                    if numb_mines > 9:
                        print('Insert maximum 9 mines')
                    else:
                        for i in range(0,numb_mines):
                            row_mine = random.randint(1,9)+1
                            col_mine = random.randint(1,9)+1
                            maze[row_mine][col_mine] = "(O)" 
                        drawMaze()
                        state = "waiting_user_input"
                except ValueError: # if inputs are not numeric
                    print('Insert a numeric value')
        
##Collect inputs (row and column numbers) to find out where do the user wants to point reflection walls: /// and \\\
##Add reflection walls to the maze as instructed by the user
            
        if user_input=='F':
            state = "adding_forward_deflector"
            while state == "adding_forward_deflector":
                try:
                    forw_defl_row = int(input('Insert the row number of the forward deflector'))
                    forw_defl_col = int(input('Insert the col number of the forward deflector'))
                    if forw_defl_row not in range(1,10) or forw_defl_col not in range(1,10):
                        print('Insert valid coordinates in the range 1-9')
                    else:
                        placeForwDefl(forw_defl_row,forw_defl_col)
                        drawMaze()
                        state = "waiting_user_input"
                except ValueError: # if inputs are not numeric
                    print('Insert numeric coordinates')
                    
        if user_input=='B':
            state = "adding_backward_deflector"
            while state == "adding_backward_deflector":
                try:
                    back_defl_row = int(input('Insert the row number of the backward deflector'))
                    back_defl_col = int(input('Insert the col number of the backward deflector'))
                    if back_defl_row not in range(1,10) or back_defl_col not in range(1,10):
                        print('Insert valid coordinates in the range 1-9')
                    else:
                        placeBackDefl(back_defl_row,back_defl_col)
                        drawMaze()
                        state = "waiting_user_input"
                except ValueError: # if inputs are not numeric
                    print('Insert numeric coordinates')
                   
        else:
            print('Select a correct option')
            drawMaze()
        
        if user_input=='S' and started == False:
            print('Start')
            started = True
            state="shooting"
        elif user_input=='S' and started == True:
            print('Game already played. Restart if you wish to replay')
            state = "waiting_user_input"
            
        while state=="shooting":
            if direction=="right":
                col_source = col_source + 1
            if direction=="up":
                row_source = row_source - 1
            if direction=="left":
                col_source = col_source - 1
            if direction=="down":
                row_source = row_source + 1    
                
            time.sleep(1) 
            clear_output(wait=True)
            drawMaze()
           
            if maze[row_source][col_source]=="///" and direction=='right':
                direction="up"
            elif maze[row_source][col_source]=="///" and direction=='up':
                direction="right"
            elif maze[row_source][col_source]=="///" and direction=='left':
                direction="down"
            elif maze[row_source][col_source]=="///" and direction=='down':
                direction="left"
            elif maze[row_source][col_source]=="\\\\\\" and direction=='right':
                direction="down"
            elif maze[row_source][col_source]=="\\\\\\" and direction=='up':
                direction="left"
            elif maze[row_source][col_source]=="\\\\\\" and direction=='left':
                direction="up"
            elif maze[row_source][col_source]=="\\\\\\" and direction=='down':
                direction="right"

            if maze[row_source][col_source]=="   ":
                maze[row_source][col_source]="+++"
            elif maze[row_source][col_source]=="=":
                print("Exit Gate Reached! Level Cleared!")
                state = "waiting_user_input"
            elif maze[row_source][col_source]=="(O)":
                print("Laser beam hits a bomb! Game Over!")
                state = "waiting_user_input"
            elif maze[row_source][col_source]=="|" or maze[row_source][col_source]=="---":
                print("Laser beam hits a wall! Game Over!")
                state = "waiting_user_input"
            elif maze[row_source][col_source]==">":
                print("Laser beam hits its source! Game Over!")
                state = "waiting_user_input" 

        if user_input=='R':
            print('Restart')
            shoot()
```
