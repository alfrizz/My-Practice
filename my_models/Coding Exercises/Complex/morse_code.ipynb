The function getMorseCode uses a technique called recursion to traverse the binary tree. When the function is called, it checks the current node. If the current nodes value is not the target character, the function calls itself on the left child and then on the right child of the current node. This process continues until it either finds the target character or reaches a leaf node (a node with no children).

When the target character is found, the function starts to unwind the recursive calls. As it unwinds, it inserts a "." or "-" into code depending on whether the character was found in the left or right subtree. This is how it builds up the Morse code for the character.

Heres a step-by-step breakdown:

Call getMorseCode(node, character, code).
If node.value is not character, call getMorseCode(node.left, character, code).
If node.left.value is not character, call getMorseCode(node.left.left, character, code).
Continue this process until node.value is character.
At this point, return True and start unwinding the recursive calls.
As each call returns, insert "." or "-" into code.
So, even though it seems like "." or "-" are only inserted when the character is found, theyre actually being inserted as the recursive calls unwind after finding the character.

```python
#Convert Character (Find the character using a pre-order traversal of the Binary Tree)
def getMorseCode(node, character, code, printvalues):
    if node==None:
        try:
            printvalues.append(node.value)
        except:
            printvalues.append('*NONE*')
        return False
    elif node.value==character:
        printvalues.append(node.value+' *FOUND*')
        return True
    else:  
        if getMorseCode(node.left, character, code, printvalues)==True:
            printvalues.append(node.value+' *LEFT*')
            code.insert(0,".")
            return True
        elif getMorseCode(node.right, character, code, printvalues)==True:
            printvalues.append(node.value+' *RIGHT*')
            code.insert(0,"-")
            return True
      
```

```python
#Message Input
message = input("Enter a message to convert into Morse Code: (e.g. SOS)").upper()
morseCode = ""

#Convert the message, one character at a time!
for character in message:
    dotsdashes = []
    printvalues = []
    getMorseCode(tree, character, dotsdashes, printvalues)
    print(printvalues)
    print('total steps:', len(printvalues))
    print('leaf none steps:', len([null_value for null_value in printvalues if null_value=='*NONE*']))
    code = "".join(dotsdashes)
    morseCode = morseCode + code + " "
  
print(morseCode)
```

```python
def decode_morse(enc_text):
    dec_text = ''
    enc_words = enc_text.split('  ')
    for enc_word in enc_words:
        enc_letts = enc_word.split()
        dec_word = ''
        for enc_lett in enc_letts:
            albero=tree
            for pos in range(len(enc_lett)):
                if enc_lett[pos] == '.':
                    albero = albero.left
                elif enc_lett[pos] == '-':
                    albero = albero.right
                else:
                    print('wrong inputs')
                    break
            dec_word += albero.value
        dec_text += ' ' + dec_word
    print(dec_text)
```
