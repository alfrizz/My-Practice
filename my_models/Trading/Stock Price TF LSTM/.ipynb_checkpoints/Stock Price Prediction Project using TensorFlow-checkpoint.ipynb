```python
ticker_name = 'AMZN'

ticker = data[data['Name'] == ticker_name]
prediction_range = ticker.loc[(ticker['date'] > datetime(2013,1,1))
 & (ticker['date']<datetime(2018,1,1))]
plt.plot(ticker['date'],ticker['close'])
plt.xlabel("Date")
plt.ylabel("Close")
plt.title(ticker_name + " Stock Prices")
plt.show()
```

Now let’s select a subset of the whole data as the training data so, that we will be left with a subset of the data for the validation part as well.

```python
close_data = ticker.filter(['close'])
dataset = close_data.values
training_len = int(np.ceil(len(dataset) * .7))

print(dataset.shape)
print(training_len)
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(dataset)
 
train_data = scaled_data[0:int(training_len), :]

train_data[0:7,:]
```

```python
n_feat = 7 # previous 'close' values to consider as features

# prepare feature and labels
x_train = []
y_train = []
 
for i in range(n_feat, training_len):
    x_train.append(train_data[i-n_feat:i]) #append a range of 'n_feat' previous 'close' values at each iteration
    y_train.append(train_data[i]) #append only last 'close' value at each iteration
    
    # x_train.append(train_data[i-n_feat:i, 0]) #append a range of 'n_feat' previous 'close' values at each iteration
    # y_train.append(train_data[i, 0]) #append only last 'close' value at each iteration

x_train, y_train = np.array(x_train), np.array(y_train)

# x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))    

for i in range(7):
    print(i)
    print(x_train[i])
    print(y_train[i])
```

```python
'''
If we want to predict 'y_train' (e.g. stock closing price) 
using as 'x_train' its 'n_feat' preceeding closing price values ('n' previous 'close' values to consider as features),
the possible number of samples to be used is: 'Data - n'
'''

from matplotlib import image as mpimg
image = mpimg.imread('Train&Test Data.jpg')
plt.imshow(image)
plt.show()
```

To stack multiple LSTM in TensorFlow it is mandatory to use return_sequences = True. 
return_sequences: Boolean. Whether to return the last output in the output sequence, or the full sequence. Default: False.
Since our data is time series varying, we apply no activation to the output layer and it remains as 1 node

```python
model = keras.models.Sequential()
model.add(keras.layers.LSTM(units=128, # 64 before
                            return_sequences=True, # to access the entire sequence of outputs from the LSTM layer
                            input_shape=(n_feat, 1))) #input_shape = (numb of timesteps: past 'close' values -7-, number of features -1-)
model.add(keras.layers.LSTM(units=128)) # 64 before
model.add(keras.layers.Dense(64)) # 32 before
model.add(keras.layers.Dropout(0.3)) # 0.5 before
model.add(keras.layers.Dense(1)) # 1 node because it's a regression problem preedicting a single continuous value
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping_monitor = EarlyStopping(monitor='loss', patience=7)

model.compile(optimizer='adam',
              metrics=[MeanSquaredError(), MeanAbsoluteError(), RootMeanSquaredError()],
              loss='mean_squared_error')
history = model.fit(x_train,
                    y_train,
                    epochs=100,
                   callbacks=[early_stopping_monitor])
```

```python
test_data = scaled_data[training_len - n_feat:, :]
test_data[0:7,:]
```

```python
y_test = dataset[training_len:, :] # 'close' values not normalized, to be compared with predictions (that will be inverse-tranformed)

x_test = []
for i in range(n_feat, len(test_data)):
    x_test.append(test_data[i-n_feat:i]) #append a range of 'n_feat' previous 'close' values at each iteration
    
    # x_test.append(test_data[i-n_feat:i, 0]) #append a range of 'n_feat' previous 'close' values at each iteration

# x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

x_test = np.array(x_test)

for i in range(7):
    print(i)
    print(x_test[i])
    print(y_test[i])
```

```python
# predict the testing data
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)
 
# evaluation metrics
mse = np.mean(((predictions - y_test) ** 2))
print("MSE", mse)
print("RMSE", np.sqrt(mse))

for i in range (7): #range(predictions.shape[0]):
    print(i,'pred:',predictions[i])
    print(i,'y_test:',y_test[i])
    print(i,'diff:',predictions[i] - y_test[i])
```
