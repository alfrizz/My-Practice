```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0
sigma = 1

normal_ds = []
uniform_ds = []

for _ in range(10000):
    normal_ds.append(np.random.normal(mu, sigma)) # Draw from the normal distribution
    uniform_ds.append(norm.cdf(normal_ds[-1], mu, sigma)) # Calculate the CDF value
```

```python
plt.hist(normal_ds, bins = 100)
```

```python
plt.hist(uniform_ds, bins = 100)
```
