```python
import numpy as np
import matplotlib.pyplot as plt

# Set the seed for reproducibility
np.random.seed(0)

# Generate a positive covariance dataset (Brown Cluster)
mean1 = [0, 0]
cov1 = [[1, 0.9], [0.9, 1]]  # diagonal elements are variances, off-diagonal elements are covariances
x1, y1 = np.random.multivariate_normal(mean1, cov1, 500).T

# Generate a negative covariance dataset (Orange Cluster)
mean2 = [0, 0]
cov2 = [[1, -0.6], [-0.6, 1]]  # diagonal elements are variances, off-diagonal elements are covariances
x2, y2 = np.random.multivariate_normal(mean2, cov2, 500).T

# Generate a zero covariance dataset (Blue Cluster)
mean3 = [0, 0]
cov3 = [[5, 0], [0, 1]]  # diagonal elements are variances, off-diagonal elements are covariances
x3, y3 = np.random.multivariate_normal(mean3, cov3, 500).T

# Plot the datasets
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.scatter(x1, y1, color='brown')
plt.text(0.5, 0.9, 'Positive(large) Covariance: \n[1, 0.9]\n[0.9, 1]', horizontalalignment='center', verticalalignment='top', transform=plt.gca().transAxes, fontsize=14)
plt.axis('equal')

plt.subplot(1, 3, 2)
plt.scatter(x2, y2, color='orange')
plt.text(0.5, 0.85, 'Negative(small) Covariance: \n[1, -0.6]\n[-0.6, 1]', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=14)
plt.axis('equal')

plt.subplot(1, 3, 3)
plt.scatter(x3, y3, color='blue')
plt.text(0.5, 0.83, 'Zero Covariance and \n x variance > y variance: \n[5, 0]\n[0, 1]', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=14)
plt.axis('equal')

plt.show()
```
