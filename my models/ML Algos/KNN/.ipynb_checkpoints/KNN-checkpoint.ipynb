```python
import numpy as np
from sklearn import preprocessing, neighbors
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import sys
import statistics
from statistics import mode
```

```python
df = pd.read_csv('breast-cancer-wisconsin.data.txt', index_col = False)
df.replace('?', -99999, inplace = True) # to treat the unknown data as outliers
df.drop('id',axis=1,inplace=True)
df = df.astype({' bare_nuclei':'int'})
```

```python
df.tail()
```

```python
X = df.drop('class', axis=1)
X.head()
```

```python
y = df['class']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
Train_set = pd.concat([X_train, y_train], axis = 1)
Test_set = pd.concat([X_test, y_test], axis = 1)
```

```python
clf= neighbors.KNeighborsClassifier(n_neighbors = 5)
clf.fit(X_train.values, y_train)
```

```python
accuracy = clf.score(X_test.values, y_test)
accuracy
```

```python
example = [[4,2,1,1,1,2,3,2,1],[4,2,1,2,1,2,3,2,1]]
prediction = clf.predict(example)
prediction
```

```python
def euclidean_distance(datapoint1, datapoint2):
    return np.round(np.sqrt(np.sum([(datapoint1[i]-datapoint2[i])**2 for i in range(len(datapoint1))])),2)
```

```python
#knn with plot for 2D features

def knn_alf_2D(dataset, feat_to_predict, k=3):
    distances = []
    if k > sum([len(i) for i in dataset.values()]):
        warnings.warn('k can''t be greater than the samples number')
        sys.exit()
    for key,value in dataset.items():
        for datapoint in value:
            plt.scatter(datapoint[0],datapoint[1], color = key)
            plt.annotate(euclidean_distance(datapoint,feat_to_predict), xy = (datapoint[0],datapoint[1]) )
            distances.append([euclidean_distance(datapoint,feat_to_predict),key])  
    predicted_class = mode([i[1] for i in sorted(distances)[:k]])
    plt.scatter(feat_to_predict[0], feat_to_predict[1], color = predicted_class)
    plt.axis('equal')
    return predicted_class, sorted(distances)
```

```python
knn_dataset = {'red':[[1,2],[2,2],[2,1],[3,1.5]],
               'green':[[4,5],[6,5],[4,6],[5,7]], 
               'yellow':[[1,9],[2,8],[1,7],[2,9.5]]}

new_feature = [1.5,5]

knn_alf_2D(knn_dataset, new_feature, 3)
```

```python
#knn for nD features

def knn_alf_nD(train_set, feat_to_predict, k=5):
    distances = []
    Train_set_dict = {2:[], 4:[]}
    
    if k > train_set.shape[0]:
        warnings.warn('k can''t be greater than the samples number')
        sys.exit()
        
    for row in range(train_set.shape[0]):
        if train_set.iloc[row]['class'] == 2:
            Train_set_dict[2].append(train_set.iloc[row][:-1].values)
        else: # class == 4
            Train_set_dict[4].append(train_set.iloc[row][:-1].values)
            
    for key,value in Train_set_dict.items():
        for datapoint in value:
            distances.append([euclidean_distance(feat_to_predict, datapoint),key])  
    top_predictions = [i[1] for i in sorted(distances)[:k]]
    predicted_class = mode(top_predictions)
    maj_class_predictions = top_predictions.count(predicted_class)
    confidence = maj_class_predictions / k
    return feat_to_predict, predicted_class, confidence
```

```python
# example = [3,4,5,3,3,3,8,4,1]

guessed = 0
total = Train_set.shape[0]
for test_sample in Train_set.values:
    feature , prediction, confidence = knn_alf_nD(Train_set, test_sample[:-1])
    actual = test_sample [-1]
    print ('feature:',feature, 'prediction:',prediction, 'actual:', actual, 'confidence:', confidence)
    if prediction == actual:
        guessed+=1
accuracy = guessed/total
print ('guessed:', guessed, 'total:', total, 'accuracy:', accuracy)
    
```
