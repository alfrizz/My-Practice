```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
style.use('ggplot')

class Support_Vector_Machine:
    
##############################################################################################     
    
    def __init__(self, visualization=True):
        print ('*** __init__ STARTED ***\n')
        self.visualization = visualization
        self.colors = {1:'r',-1:'b'}
        if self.visualization:
            self.fig = plt.figure()
            self.ax = self.fig.add_subplot(1,1,1)
        print ('*** __init__ ENDED ***\n')

##############################################################################################          
            
    # train
    def fit(self, data_dict):
        print ('*** fit STARTED ***\n')        
        self.data = data
        print('data:\n',self.data)
        
        # to store {norm: w_t, b} for each iteration
        opt_dict = {} 
        
        # to transform w to w_t, to get all combinations of positive and negative values (4 in a 2D space)
        transforms = [[1,1],
                      [-1,1],
                      [1,-1],
                      [-1,-1]] 
        
        self.max_feature_value = np.max([i for i in data_dict.values()])
        print ('max_feature_value:',self.max_feature_value)
        self.min_feature_value = np.min([i for i in data_dict.values()])
        print ('min_feature_value:',self.min_feature_value,'\n')

        
        step_sizes = [self.max_feature_value * 0.1,
                      self.max_feature_value * 0.01,
                      self.max_feature_value * 0.001,
                      # self.max_feature_value * 0.0001, # expensive
                      ]   
        print ('step_sizes:',step_sizes,'\n')
        
        # 'w' must be better tuned than 'b', so we use a bigger multiplier (we dont need to take as small of steps with 'b', as we do with 'w')
        w_range_multiple = 10
        b_range_multiple = 3
        
        b_multiple = 3 # multipliers for b steps (in this case, the lower is the value, the slower is the execution)

        latest_optimum = self.max_feature_value*w_range_multiple
        print ('latest_optimum:',self.max_feature_value*w_range_multiple,'\n') # we define the highest value, as starting point for the optimization
        
        step_num = 1
        for step in step_sizes:
            w = np.array([latest_optimum,latest_optimum]) # highest values of 'w', as starting point for its minimization (num. w = num. features)
            # we can do this because convex
            optimized = False
            while not optimized:
                for b in np.arange(-1*(self.max_feature_value*b_range_multiple),
                                   self.max_feature_value*b_range_multiple,
                                   step*b_multiple):
                    for transformation in transforms: # only one of the 4 transformations will be used, accordingly to the sample data distribution
                        w_t = w*transformation  # to transform w to w_t, to get all combinations of positive and negative values (4 in a 2D space)
                        found_option = True
                        for yi in self.data:
                            for xi in self.data[yi]:
                                if not yi*(np.dot(w_t,xi)+b) >= 1: # first constraint check: yi(xi.w+b) >= 1 (must by complied by all samples)
                                    found_option = False
                                    break
                            if not found_option:
                                break
                        if found_option:
                            opt_dict[np.linalg.norm(w_t)] = [w_t,b]
                            
                            # print ('yi:',yi)
                            # print ('xi:',xi)
                            # print ('step:',step)
                            # print ('w:',w)
                            # print ('w[0]:',w[0])
                            # print ('transformation:',transformation)
                            # print ('w_t:',w_t)
                            # print ('b:',b)
                            # print ('norm:',np.linalg.norm(w_t))
                            # print ('yi*(xi*w_t + b) >= 1:',yi*(np.dot(w_t,xi)+b))
                            # print ('*******************************************')
                            
                if w[0] < 0:
                    optimized = True
                    print('\033[31m Optimized step:', step_num,'\033[0m\n')
                    print ('\033[1m last ''w'' negative that interrupts the loop: :\033[0m\n',w[0])
                    step_num += 1
                else:
                    w = w - step                  
            # print ('\033[1m opt_dict {||w||: w_t,b}: \033[0m\n',opt_dict,'\n')
            norms = sorted([n for n in opt_dict])
            opt_choice = opt_dict[norms[0]] #we are interested in the values associated to the minimum norm   
            print ('\033[1m opt_choice (w_t and b, for the minimum norm) :\033[0m\n',opt_choice,'\n')
            self.w = opt_choice[0] #taking the 'w_t' associated to the minimum norm
            print ('self.w = opt_choice[0] (optimal w_t):',opt_choice[0],'\n')
            self.b = opt_choice[1] #taking the 'b' associated to the minimum norm
            print ('self.b = opt_choice[1] (optimal b):',opt_choice[1],'\n')
            latest_optimum = np.abs(opt_choice[0][0]) # +step
            print ('latest_optimum ( last w_t, to be reused as new ''w'' value = opt_choice[0][0])',latest_optimum,'\n')
            
        for yi in self.data:
            for xi in self.data[yi]:
                print ('********************* SAMPLES DATA POINTS **************************')
                print ('xi:',xi)
                print ('yi:',yi)
                print ('self.w:',self.w)
                print ('self.b:',self.b)
                print ('xi*w + b:',np.dot(self.w,xi)+self.b)
                print('xi: yi*(xi*w + b) <--->',xi,':',yi*(np.dot(self.w,xi)+self.b))  
        print ('\n*** fit ENDED ***\n')

##############################################################################################                 
                
    def predict(self,features):
        print ('*** predict STARTED ***')
        # sign( x.w+b )
        classification = np.sign(np.dot(np.array(features),self.w)+self.b)
        if classification !=0 and self.visualization:
            self.ax.scatter(features[0], features[1], s=200, marker='*', c=self.colors[classification])
        return classification
        print ('*** predict ENDED ***')

##############################################################################################     
    
    def visualize(self):
        print ('*** visualize STARTED ***')
        [[self.ax.scatter(x[0],x[1],s=100,color=self.colors[i]) for x in data_dict[i]] for i in data_dict]

        # x*w+b = v = (1 high boundary, -1 low boundary, 0 decision boundary)
        def hyperplane(x,w,b,v):
            return (-w[0]*x-b+v) / w[1] # formula  to compute the y-coordinate of the point where the line intersects the x-axis, 
                                        # which is the distance between the data point and the decision boundary.

        #multiplications to 0.9 and 1.1 to add some margin, besides the features
        hyp_x_min = self.min_feature_value*0.9 # min 'x' range
        hyp_x_max = self.max_feature_value*1.1 # max 'x' range

        # (w.x+b) = 1
        # positive support vector hyperplane
        psv1 = hyperplane(hyp_x_min, self.w, self.b, 1)
        psv2 = hyperplane(hyp_x_max, self.w, self.b, 1)
        self.ax.plot([hyp_x_min,hyp_x_max],[psv1,psv2], 'k')

        # (w.x+b) = -1
        # negative support vector hyperplane
        nsv1 = hyperplane(hyp_x_min, self.w, self.b, -1)
        nsv2 = hyperplane(hyp_x_max, self.w, self.b, -1)
        self.ax.plot([hyp_x_min,hyp_x_max],[nsv1,nsv2], 'k')

        # (w.x+b) = 0
        # positive support vector hyperplane
        db1 = hyperplane(hyp_x_min, self.w, self.b, 0)
        db2 = hyperplane(hyp_x_max, self.w, self.b, 0)
        self.ax.plot([hyp_x_min,hyp_x_max],[db1,db2], 'y--')

        plt.show()
        print ('*** visualize ENDED ***')
        print (self.min_feature_value,self.max_feature_value)
        
```
