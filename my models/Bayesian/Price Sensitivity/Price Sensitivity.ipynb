```python
from sklearn.model_selection import train_test_split
```

```python
X = clients_df[['price', 'country']]
y = clients_df['purchase']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# The model below tries to learn the relationship between the price of a product and the probability of a purchase, taking into account the uncertainty in the parameters. The output is a distribution of plausible values for b0 and b1 given the data, rather than single point estimates. This allows for a more nuanced understanding of the parameter values and their uncertainty.

with pm.Model() as pooled_model:

    # define b0 intercept and b1 parameter priors (the standard deviations below are quite large (100), which means weâ€™re actually quite uncertain about the value of the coefficient and intercept)
    b0 = pm.Normal('b0', mu=0, sigma=100) #  A normal distribution for b0 implies that we believe the intercept could be any real number, with a higher probability around 0
    b1 = pm.Lognormal('b1', mu=0, sigma=100) # A lognormal distribution for b1 implies that we believe the price coefficient is positive and most likely close to 1 (since the log of 1 is 0), but could potentially be much larger. This could make sense if we believe that price has a multiplicative effect on the odds of purchase (the effect is inverse, hence the '-' sign in the logit_p formula below).

    # compute the purchase probability for each case
    logit_p = b0 - b1 * X_train['price'].to_list()
    likelihood = pm.Bernoulli('likelihood', logit_p=logit_p, observed=list(y_train)) # In this case, each row in the data is considered a single trial, with â€˜purchaseâ€™ as the success. Therefore, a Bernoulli distribution is appropriate for the likelihood. If you had data on multiple trials for each row (like the number of purchases out of a number of offers), then a binomial distribution would be appropriate.
    
    posterior = pm.sample(draws = 6000, tune = 3000) # The posterior distribution, which is the updated belief about the parameters after seeing the data, is sampled using a Markov Chain Monte Carlo (MCMC) method. The pm.sample() function is used to draw 6000 samples after a tuning phase of 3000 iterations.
    
```

```python
# Hierarchal Bayesian model
n_countries = X_train['country'].nunique()
n_countries
```

```python
country_to_index = {'Italy' : 0, 'Germany' : 1, 'France' : 2}
ids_train = X_train['country'].apply(lambda x: country_to_index[x])
ids_test = X_test['country'].apply(lambda x: country_to_index[x])
```

```python
ids_train
```

```python
with pm.Model() as hier_model:
    
# the parameters b0 and b1 are assumed to vary across different countries. By placing a distribution (the Normal distribution for mu_b0 and mu_b1, and the Inverse Gamma distribution for sigma_b0 and sigma_b1) on the parameters of the b0 and b1 distributions, weâ€™re saying that we believe these parameters come from a common distribution, but weâ€™re uncertain about the exact parameters of this common distribution.
    
    # priors for common distribution's mu and sigma
    mu_b0 = pm.Normal('mu_b0', mu=0, sigma=100)
    mu_b1 = pm.Normal('mu_b1', mu=0, sigma=100)

    sigma_b0 = pm.InverseGamma('sigma_b0', alpha=3, beta=20)
    sigma_b1 = pm.InverseGamma('sigma_b1', alpha=3, beta=20)

# In the hierarchical model, we are creating three separate distributions for the parameters b0 and b1, one for each country(shape=n_countries). These distributions are not necessarily equal, but they are generated from the same type of distribution (Normal for b0 and Log-normal for b1) with hyperparameters (mu_b0, mu_b1, sigma_b0, sigma_b1) that are common across all countries.
    
    # country-specific draws 
    b0 = pm.Normal('b0', mu=mu_b0, sigma=sigma_b0, shape=n_countries)
    b1 = pm.Lognormal('b1', mu=mu_b1, sigma=sigma_b1, shape=n_countries)

# Then, for each observation in the training data, we are calculating a logit value that is specific to the country of that observation. We do this by selecting the b0 and b1 values that correspond to the country of the observation (b0[ids_train] and b1[ids_train]), and then using these values to calculate the logit. Each logit_p value is calculated using the b0 and b1 values that correspond to the country of each observation
    
    # compute the purchase probability for each case
    logit_p = b0[ids_train] - b1[ids_train] * list(X_train['price'])
    likelihood = pm.Bernoulli('likelihood', logit_p=logit_p, observed=list(y_train))
    
    posterior = pm.sample(draws = 10000, tune = 5000, target_accept = 0.9)
```

```python
def make_hier_prediction(prices: np.array, country_index: int, posterior_values: np.array = posterior, get_whole_distribution: bool = False) -> float:
    
    """
    Given an array of prices (or a single)
    """

# makes the prediction using the mean of the parameters of the country selected
    
    country_b0 = posterior['posterior']['b0'][0].values[:, country_index]
    country_b1 = posterior['posterior']['b1'][0].values[:, country_index]
    
    if not get_whole_distribution:
        country_b0 = country_b0.mean()
        country_b1 = country_b1.mean()
    
# the format of logit_value will be different depending on the value of get_whole_distribution.

# If get_whole_distribution is False, country_b0 and country_b1 are scalars (the mean values), so logit_value will also be a scalar. In this case, predicted_probability will be a scalar.
# If get_whole_distribution is True, country_b0 and country_b1 are arrays representing the entire distributions, so logit_value will be an array. In this case, predicted_probability will also be an array, with each element representing the predicted probability for a corresponding logit_value.

# The np.exp(logit_value) / (1 + np.exp(logit_value)) calculation can handle both scalar and array inputs, thanks to the broadcasting feature of NumPy    
    
    logit_value = country_b0 - country_b1 * prices
    predicted_probability = np.exp(logit_value) / (1 + np.exp(logit_value))
    
    return predicted_probability
```

```python
# getting test predictions

y_test_predict = []
for i in range(len(X_test)):
    country_idx = country_to_index[X_test['country'].iloc[i]]
    predicted_prob = make_hier_prediction(X_test['price'].iloc[i], country_idx)
    y_test_predict.append(predicted_prob >= 0.5)
```

```python
# Now that we know price sensitivity in each country, we can compare various pricing strategies. Letâ€™s say Virtuoso wants to set a fixed price of 7 for all clients and offer a (temporary) discount of 2 for (more price-sensitive) clients in Germany. We can calculate not just point estimates but whole probability distributions for price acceptance of clients in each country.
```

```python
italy_prob7 = make_hier_prediction(np.array([7]), 0, get_whole_distribution = True)
germany_prob5 = make_hier_prediction(np.array([5]), 1, get_whole_distribution = True)
france_prob7 = make_hier_prediction(np.array([7]), 2, get_whole_distribution = True)
```
