```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.cluster import MeanShift, KMeans, estimate_bandwidth
from sklearn.datasets import make_blobs
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import style
from operator import itemgetter
# style.use("ggplot")

centers = [[1, 1, 1], [5, 5, 5], [3, 10, 10]]

X, _ = make_blobs(n_samples = 100, centers = centers, cluster_std = 1)

ms = MeanShift()
ms.fit(X)
labels = ms.labels_
cluster_centers = ms.cluster_centers_

print(cluster_centers)
n_clusters_ = len(np.unique(labels))
print("Number of estimated clusters:", n_clusters_)

colors = 10*['r','g','b','c','k','y','m']
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

print (X.shape)

for i in range(len(X)):
    ax.scatter(X[i][0], X[i][1], X[i][2], c=colors[labels[i]], marker='o') # x,y,z coordinates for the data blob

ax.scatter(cluster_centers[:,0],cluster_centers[:,1],cluster_centers[:,2], # x,y,z coordinates for the cluster centers
            marker="x",color='k', s=150, linewidths = 5, zorder=10)

plt.show()
```

TITANIC DATASET

```python
'''
Pclass Passenger Class (1 = 1st; 2 = 2nd; 3 = 3rd)
survival Survival (0 = No; 1 = Yes)
name Name
sex Sex
age Age
sibsp Number of Siblings/Spouses Aboard
parch Number of Parents/Children Aboard
ticket Ticket Number
fare Passenger Fare (British pound)
cabin Cabin
embarked Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)
boat Lifeboat
body Body Identification Number
home.dest Home/Destination
'''

df = pd.read_excel('titanic.xls')
df.drop(['body','name','ticket'], axis=1, inplace = True)
df['fare'].fillna(0, inplace=True)
df['fare'] = df['fare'].astype(int)
df['age'].fillna(np.mean(df['age']), inplace=True)

df.cabin.fillna('Z', inplace=True)
df['cabin_lett'] = [df.cabin[i][0] for i in range(df.shape[0])] 
df.drop('cabin', axis=1, inplace = True)

#get_dummies
col_to_get_dum = ['sex', 'embarked', 'cabin_lett']
dummies = pd.get_dummies(df[col_to_get_dum], drop_first = True)
df = pd.concat([df, dummies], axis=1)
df.drop(col_to_get_dum, axis=1, inplace = True)

#label_encoders
col_to_lab_enc = ['boat']
le = LabelEncoder()
for i in range(len(col_to_lab_enc)):
    df[col_to_lab_enc[0]+'_labelencod'] = le.fit_transform(df[col_to_lab_enc].astype(str))
    df.drop(col_to_lab_enc, axis=1, inplace = True)

#top feature_values encoding
col_to_top_feat = ['home.dest']
for i in range(len(col_to_top_feat)):
    for j in range(10): # top 10 feature_values  
        df.loc[df[col_to_top_feat[i]] == df[col_to_top_feat[i]].value_counts().index[j], col_to_top_feat[i]+'_topfeatval'] = j+1
    df[col_to_top_feat[i]+'_topfeatval'].fillna(0, inplace=True)
    df.drop(col_to_top_feat, axis=1, inplace = True)
    
X = df.drop('survived', axis=1)
y_true = df.survived

scaler = MinMaxScaler()
# fit the scaler to the data and transform the data
X_norm = scaler.fit_transform(X)

X.head()
```

```python
bandwidth = estimate_bandwidth(X_norm, quantile=0.5)
clf = MeanShift(bandwidth = bandwidth)
clf.fit(X_norm)
```

```python
labels = clf.labels_
cluster_centers = clf.cluster_centers_
df['cluster_group']=labels
df.head()
```

MEAN SHIFT FROM SCRATCH

```python
class Mean_Shift_Alf:
    
    def __init__(self, radius = 1, max_iter = 10, tol = 0.001):
        self.max_iter = max_iter
        self.radius = radius
        self.tol = tol
                
        
        
    def fit(self, dataset, print_distances = False):
        '''
        When a data point is within the bandwidth of multiple centroids,
        the data point is then assigned to the cluster with the highest density, 
        which corresponds to the cluster with the most data points.
        '''
        self.centroids = {}
        self.in_bandwidth = {}
        self.matching_centroids = {}
        self.final_clusters = {}
        centroids_distance = {}
        centr_dist_diff = {}
        
        centroids_0 = [datapoint for datapoint in dataset] # a centroid is assigned to every datapoint
        
        for iteration in range(self.max_iter):
            #list of all centroids (for each iteration - outer dictionary)
            self.centroids[iteration] = np.unique([np.round(np.mean(dps_in_bw, axis=0),2)
                                        for _ , dps_in_bw in self.in_bandwidth[iteration - 1].items()], axis = 0).tolist()\
                                        if iteration > 0 else centroids_0   
            
            self.in_bandwidth[iteration] = {}
            self.matching_centroids[iteration] = {}
            

            centroids_distance[iteration] = [np.linalg.norm(np.array(self.centroids[iteration][n_cluster]) - 
                                                            np.array(self.centroids[iteration-1][n_cluster])) 
                                             if (iteration > 0 
                                                 and len(self.centroids[iteration]) == len(self.centroids[iteration -1]))
                                             else None
                                             for n_cluster in range(len(self.centroids[iteration]))] 

            centr_dist_diff[iteration] = [np.abs((np.array(centroids_distance[iteration][n_cluster]) 
                                                  - np.array(centroids_distance[iteration - 1][n_cluster]))
                                                 / np.array(centroids_distance[iteration - 1][n_cluster]))
                                          if (iteration > 1
                                              and centroids_distance[iteration][n_cluster] is not None
                                              and centroids_distance[iteration - 1][n_cluster] is not None
                                              and centroids_distance[iteration][n_cluster]!=0 
                                              and centroids_distance[iteration - 1][n_cluster]!=0)
                                          else 0 
                                          if (iteration > 1
                                              and centroids_distance[iteration][n_cluster] is not None
                                              and centroids_distance[iteration - 1][n_cluster] is not None
                                              and (centroids_distance[iteration][n_cluster]==0 
                                              or centroids_distance[iteration - 1][n_cluster]==0))
                                          else None
                                          for n_cluster in range(len(self.centroids[iteration]))]
            
            if print_distances:
                print('\niteration', iteration, 
                      '\ncentroid', self.centroids[iteration],
                     '\ndistance', centroids_distance[iteration],
                     '\ndist_diff', centr_dist_diff[iteration])
            
            if all(val is not None 
                   and val < self.tol 
                   for val in centr_dist_diff[iteration]):
                self.max_iter = iteration
                break   
                
                
            
            for centroid in self.centroids[iteration]:
                #dictionary of all centroids with all datapoints within their bandwidth (for each iteration - outer dictionary)
                self.in_bandwidth[iteration][tuple(centroid)] = [dp_in_bw for dp_in_bw in dataset 
                                                            if np.linalg.norm(dp_in_bw - centroid) < self.radius] 
                
                #dictionary of all datapoints with the candidate centroids and their density: number of containing datapoints
                #(for each iteration - outer dictionary)
                for datapoint in dataset:
                    self.matching_centroids[iteration][tuple(datapoint)] = [(cand_centroids, len(self.in_bandwidth[iteration][tuple(cand_centroids)])) 
                                                                            for cand_centroids in self.in_bandwidth[iteration]
                                                                            if any([np.array_equal(datapoint, dp_in_bw)
                                                                            for dp_in_bw in self.in_bandwidth[iteration][tuple(cand_centroids)]])]
            
            #dictionary of all datapoints with the related centroid associated to the most dense cluster: 
            #(for each iteration - outer dictionary)
            self.final_clusters[iteration] = dict([(datapoint, max(cand_centroids, key=lambda x: x[1])[0]) 
                                                   #x[1]:number of datapoints, x[0]:centroid
                                                   for datapoint, cand_centroids in self.matching_centroids[iteration].items()
                                                   if cand_centroids])                    
                    

    
    def plot(self, dataset, print_centroids = False):
        colors = ['red', 'green', 'blue', 'cyan', 'magenta', 'yellow', 'orange', 'pink', 'purple', 'brown', 'maroon', 'navy', 'teal', 'olive', 'gray']
        dict_centr_col = {} #dictionary of centroids and associated colors
        
        for iteration in range(self.max_iter):
            _, ax = plt.subplots()
            ax.axis('equal')
            plt.xlim(min(col[0] for col in self.centroids[iteration]) - 2*self.radius, max(col[0] for col in self.centroids[iteration]) + 2*self.radius)
            plt.ylim(min(col[1] for col in self.centroids[iteration]) - 2*self.radius, max(col[1] for col in self.centroids[iteration]) + 2*self.radius)
            
            colors = int(np.round(len(self.centroids[iteration])/len(colors))+1)*colors #to replicate the colors list n times, accordig to the number of centroids to plot

            for i in range(len(self.centroids[iteration])):
        
                ax.scatter(self.centroids[iteration][i][0], 
                           self.centroids[iteration][i][1],
                           marker = '*', c = colors[i], s = 100, alpha = 0.3)

                ax.annotate(self.centroids[iteration][i] , xy = (self.centroids[iteration][i][0],
                                                                 self.centroids[iteration][i][1]+0.2), 
                            fontsize=8, c = colors[i], alpha = 0.3)
                
                circle = plt.Circle((self.centroids[iteration][i][0], self.centroids[iteration][i][1]), 
                                    self.radius, fill=False, color = colors[i], alpha = 0.3)
                ax.add_artist(circle)
                
                dict_centr_col[tuple(self.centroids[iteration][i])] = colors[i]
                
            
            for tuple_dp_cen in self.final_clusters[iteration].items():
                datapoint, centroid = tuple_dp_cen
                ax.scatter(datapoint[0], datapoint[1],  c = dict_centr_col[centroid])
                
            plt.show()
        
            if print_centroids:
                       print ('\ncentroids', self.centroids[iteration], 
                       '\nin_bandwidth', self.in_bandwidth[iteration],
                       '\nmatching_centroids', self.matching_centroids[iteration],
                       '\nfinal_clusters', self.final_clusters[iteration])
        
        
           

                
```
