```python
import pandas as pd 
import numpy as np 
import torch.nn as nn
from pytorch_pretrained_bert import BertTokenizer, BertModel
import torch
# from keras.preprocessing.sequence import pad_sequences
from keras_preprocessing.sequence import pad_sequences
from sklearn import tree
from sklearn.metrics import classification_report
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv("fake_job_postings.csv")
print(df.head())
```

```python
df = df[['description', 'fraudulent']]
print(df.head())
```

```python
from collections import Counter
print(Counter(df['fraudulent'].values))
```

```python
df.fraudulent.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
# Next we want to balance our data set such that we have an equal number of fraudulent and not fraudulent types. 

df_fraudulent= df[df['fraudulent'] == 1] 
df_normal = df[df['fraudulent'] == 0] 
df_normal = df_normal.sample(n=len(df_fraudulent))
df = pd.concat([df_normal, df_fraudulent])
```

```python
df.fraudulent.value_counts()
```

```python
# We also should randomly shuffle the targets, to ensure that your models performance is not affected by the order of the data points. 
df = df.sample(frac=1, random_state = 24).reset_index(drop=True)
df.shape
```

```python
train_data = df.head(int(df.shape[0]/2))
test_data = df.tail(int(df.shape[0]/2)).reset_index(drop=True)
train_data.shape, test_data.shape
```

```python
train_data_dics = [{'description': train_data.loc[i, 'description'], 'fraudulent': train_data.loc[i, 'fraudulent'] } for i in range(train_data.shape[0])]
train_data_dics[0:3]
```

```python
test_data_dics = [{'description': test_data.loc[i, 'description'], 'fraudulent': test_data.loc[i, 'fraudulent'] } for i in range(test_data.shape[0])]
test_data_dics[0:3]
```

```python
# Generate a list of tuples from the list of dictionaries :

# map(lambda d: (d['description'], d['fraudulent']), train_data_dics): This applies a lambda function to each dictionary d in train_data. The lambda function returns a tuple containing the description and fraudulent fields from d.

list(map(lambda d: (d['description'], d['fraudulent']), train_data_dics))[0]
```

```python
list(zip(*map(lambda d: (d['description'], d['fraudulent']), train_data_dics)))[0][0]
```

```python
list(zip(*map(lambda d: (d['description'], d['fraudulent']), train_data_dics)))[1][0]
```

```python
train_texts, train_labels = list(zip(*map(lambda d: (d['description'], d['fraudulent']), train_data_dics)))
len(train_texts), len(train_labels)
```

```python
train_texts[0]
```

```python
train_labels[0]
```

```python
test_texts, test_labels = list(zip(*map(lambda d: (d['description'], d['fraudulent']), test_data_dics)))
len(test_texts), len(test_labels)
```

```python
# Generate tokens and token ids: Notice we truncate the input strings to 512 characters because that is the maximum number of tokens BERT can handle.
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)
tokenizer
```

```python
train_tokens = list(map(lambda t: ['[CLS]'] + tokenizer.tokenize(t)[:511], train_texts))
train_tokens[0]
```

```python
test_tokens = list(map(lambda t: ['[CLS]'] + tokenizer.tokenize(t)[:511], test_texts))
```

```python
train_tokens_ids = list(map(tokenizer.convert_tokens_to_ids, train_tokens))
train_tokens_ids[0]
```

```python
test_tokens_ids = list(map(tokenizer.convert_tokens_to_ids, test_tokens))
```

```python
# So, the pad_sequences function in your code is ensuring that all sequences in train_tokens_ids have a length of 512, by either padding shorter sequences with zeros at the end, or truncating longer sequences from the end, and the resulting sequences will be of integer type.

train_tokens_ids = pad_sequences(train_tokens_ids, maxlen=512, truncating="post", padding="post", dtype="int")
train_tokens_ids
```

```python
test_tokens_ids = pad_sequences(test_tokens_ids, maxlen=512, truncating="post", padding="post", dtype="int")
test_tokens_ids
```

```python
# Finally, generate a boolean array based on the value of fraudulent for our testing and training sets:

train_y = np.array(train_labels) == 1
train_y
```

```python
test_y = np.array(test_labels) == 1
```

```python
# We create our BERT classifier which contains an initialization method and a forward method that returns token probabilities:

# import torch.nn as nn
# This module includes classes for different types of layers in a neural network (like linear layers, convolutional layers, etc.), loss functions (like Mean Squared Error, Cross Entropy Loss, etc.), and various other utilities

class BertBinaryClassifier(nn.Module): # is a binary classifier that inherits from the nn.Module class in PyTorch
    
    def __init__(self, dropout=0.1): # constructor for the class, initializes an instance of the class and sets up the layers of the neural network
        super(BertBinaryClassifier, self).__init__() # calling constructor of parent class nn.Module. Necessary for the PyTorch internals to work
        self.bert = BertModel.from_pretrained('bert-base-uncased') # is loading a pre-trained BERT model 'bert-base-uncased'
        self.dropout = nn.Dropout(dropout) # randomly sets a fraction dropout of the input units to 0 at each update during training time
        self.linear = nn.Linear(768, 1) # linear layer. The numbers 768 and 1 are the input and output sizes respectively
        self.sigmoid = nn.Sigmoid() # sigmoid activation function useful for binary classification
    
    def forward(self, tokens, masks=None): # defines the computation performed at every call
        _, pooled_output = self.bert(tokens, attention_mask=masks, output_all_encoded_layers=False) #  feeds the input data (tokens and masks) through the BERT model.
        dropout_output = self.dropout(pooled_output) # output from the BERT model is then passed through a dropout layer
        linear_output = self.linear(dropout_output) # output from the dropout layer is passed through a linear layer
        proba = self.sigmoid(linear_output) # output from the linear layer is passed through a sigmoid activation function
        return proba # returns the probabilities computed by the sigmoid function
```

```python
# Next, we generate training and testing masks:  the purpose of this code is to create attention masks that indicate where the actual tokens are in each sequence-  If a token is present, the mask has a value of 1.0 , 0.0 otherwise

train_masks = [[float(i > 0) for i in ii] for ii in train_tokens_ids]
len(train_masks)
```

```python
#  converting the 2D lists of masks into PyTorch tensors. This is necessary because PyTorch models require inputs to be tensors.

train_masks_tensor = torch.tensor(train_masks)
train_masks_tensor
```

```python
test_masks = [[float(i > 0) for i in ii] for ii in test_tokens_ids]
test_masks_tensor = torch.tensor(test_masks)
test_masks_tensor
```

```python
# Generate token tensors for training and testing: converting the tokenized training and testing data (train_tokens_ids and test_tokens_ids) into PyTorch tensors

train_tokens_tensor = torch.tensor(train_tokens_ids)
test_tokens_tensor = torch.tensor(test_tokens_ids)
train_tokens_tensor
```

```python
#  converting the training and testing labels (train_y and test_y) into PyTorch tensors.

# The reshape(-1, 1) function call is reshaping the labels into a 2D tensor with one column. The .float() function call is ensuring that the labels are floating-point numbers, which is the expected data type for many loss functions in PyTorch.

train_y_tensor = torch.tensor(train_y.reshape(-1, 1)).float()
test_y_tensor = torch.tensor(test_y.reshape(-1, 1)).float()
train_y_tensor[0:10]
```

```python
# and finally, prepare our data loaders:

BATCH_SIZE = 5 

train_dataset =  torch.utils.data.TensorDataset(train_tokens_tensor, train_masks_tensor, train_y_tensor)
train_dataset
```

```python
train_sampler =  torch.utils.data.RandomSampler(train_dataset)
train_sampler
```

```python
train_dataloader =  torch.utils.data.DataLoader(train_dataset, sampler=train_sampler, batch_size=BATCH_SIZE)
train_dataloader
```

```python
test_dataset =  torch.utils.data.TensorDataset(test_tokens_tensor, test_masks_tensor, test_y_tensor)
test_sampler =  torch.utils.data.SequentialSampler(test_dataset)
test_dataloader =  torch.utils.data.DataLoader(test_dataset, sampler=test_sampler, batch_size=BATCH_SIZE)
```

```python
for batch_data in train_dataloader:
    for i in batch_data:
        print(i)
        print('********************************************************************************************')
    break
```

```python
# 5. FINE TUNING

# We use the Adam optimizer to minimize the Binary Cross Entropy loss and we train with a batch size of 5 for 1 EPOCHS: 

EPOCHS = 1
bert_clf = BertBinaryClassifier()
optimizer = torch.optim.Adam(bert_clf.parameters(), lr=3e-6) # the Adam optimizer will update the model parameters (weights and biases) when the optimizer.step() method is called during the training process
for epoch_num in range(EPOCHS):
    bert_clf.train() # It sets the model to training mode.
    train_loss = 0 # It initializes the training loss to 0.
    for step_num, batch_data in enumerate(train_dataloader): # It iterates over the training data (865/5 = 173 iterations)
        token_ids, masks, labels = tuple(t for t in batch_data) # It unpacks the batch data into tokens, masks, and labels.
        probas = bert_clf(token_ids, masks) # It feeds the tokens and masks into the model to get the predicted probabilities.
        loss_func = nn.BCELoss() # It calculates the batch loss using Binary Cross Entropy (BCE) loss.
        batch_loss = loss_func(probas, labels)
        train_loss += batch_loss.item() # It adds the batch loss to the total training loss.
        bert_clf.zero_grad() # resets the gradients of the model parameters to zero. Gradients in PyTorch accumulate by default
        batch_loss.backward() # computes the gradient of the loss with respect to the model parameters.
        optimizer.step() # updates the model parameters based on the gradients computed during batch_loss.backward()
        print('Epoch: ', epoch_num + 1)
        print("\r" + "{0}/{1} loss: {2} ".format(step_num, len(train_data) / BATCH_SIZE, train_loss / (step_num + 1)))
```

```python
# And we evaluate our model:

bert_clf.eval() # sets the model to evaluation mode. This is important because certain layers like dropout and batch normalization behave differently during training and evaluation
bert_predicted = [] # model predictions (0 or 1)
all_logits = [] # raw output values (percentages)
test_loss = 0
with torch.no_grad(): # context manager is used to disable gradient calculations during the evaluation phase, which can save memory and speed up computations
    for step_num, batch_data in enumerate(test_dataloader): # for each batch
        token_ids, masks, labels = tuple(t for t in batch_data) # unpacks the batch data into tokens, masks, and labels.
        logits = bert_clf(token_ids, masks) # feeds the tokens and masks into the model to get the logits
        loss_func = nn.BCELoss()
        loss = loss_func(logits, labels) # calculates the loss between the logits and the actual labels using Binary Cross Entropy (BCE)
        test_loss += loss.item()
        numpy_logits = logits.cpu().detach().numpy() # It detaches the logits from the computation graph, moves them to the CPU, and converts them to a NumPy array. With .detach() on a tensor, youre creating a tensor that shares the same data but does not require gradient computation
        bert_predicted += list(numpy_logits[:, 0] > 0.5)
        all_logits += list(numpy_logits[:, 0])
        print('Iteration', step_num+1, 'over', len(test_data)/BATCH_SIZE, 'Test loss:', test_loss/(step_num+1))
```

```python
print(classification_report(test_y, bert_predicted))
```
