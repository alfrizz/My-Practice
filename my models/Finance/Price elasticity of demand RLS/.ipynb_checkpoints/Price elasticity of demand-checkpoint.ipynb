```python
# from __future__ import print_function
import numpy as np
import pandas as pd
from pandas.tseries.offsets import *
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.compat import lzip
from statsmodels.formula.api import ols
```

```python
beef = pd.read_csv('beef.csv')
beef
```

Price elasticity of demand (PED) is a measure used in economics to show the responsiveness, or elasticity, of the quantity demanded of a good or service to a change in its price when nothing but the price changes. More precisely, it gives the percentage change in quantity demanded in response to a one percent change in price.

We will build a linear regression model to estimate PED, and we will use Python’s Statsmodels to estimate our models as well as conduct statistical tests, and data exploration.

```python
# Regression Analysis
# Ordinary Least Squares (OLS) Estimation

beef_model = ols("Quantity ~ Price", data=beef).fit()
beef_model
```

```python
beef_model.summary()

# The results indicate that the model is significant and explains a large portion of the variability in Quantity. 
# The Price coefficient is significant and suggests that as Price increases, Quantity decreases. 
# The confidence intervals for the coefficients suggest that the estimates are precise.
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_partregress_grid(beef_model, fig=fig)


# Partial Regression Plot Interpretation:

# Y-Axis Residuals:
# Definition: The residuals of the target variable (e.g., Quantity).
# Calculation:
# Residual (Y-axis) = Observed Quantity − Predicted Quantity

# X-Axis Residuals:
# Definition: The residuals of the predictor variable (e.g., Price), adjusted for the effects of other variables in the model.
# Simple Case with One Predictor: Residuals are the differences between observed values and their mean.
# General Case: If there are other predictors, the residuals are calculated by regressing the predictor on the other predictors 
# and taking the difference between the observed and predicted values.

# Intercept plot
# The plot likely shows a flat line, suggesting the intercept is stable and contributes consistently to the baseline Quantity.
# The intercept's contribution is significant in providing a starting value for the dependent variable, Quantity, in the regression model.

# Price plot
# The negative slope of the fitted line suggests that as Price increases, Quantity decreases, even after adjusting for the intercept. 
# This confirms the expected inverse relationship between Price and Quantity.
```

```python
fig = plt.figure(figsize=(12, 8))
fig = sm.graphics.plot_ccpr_grid(beef_model, fig=fig)

# Component-Component plus Residual (CCPR) Plots

# The CCPR plot provides a way to judge the effect of one regressor on the response variable by taking into account the effects of the other independent variables.
# As you can see the relationship between the variation in Quantity explained by Price is definite linear. 
# There are not many observations that are exerting considerable influence on the relationships.

# the Y-axis is the sum of 𝛽1×Price and the residuals. This means it shows how much of the Quantity is explained by Price and what's left as unexplained (residuals).
# This plot allows you to see the relationship between Price and Quantity while factoring in the intercept's contribution separately
```

```python
# Regression plots
# We use plot_regress_exog function to quickly check model assumptions with respect to a single regressor, Price in this case.

fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(beef_model, 'Price', fig=fig)
```

```python
# Recursive Least Square (RLS)

# Finally we apply Recursive Least Square (RLS) filter to investigate parameter instability.

# Before RLS estimation, we will manipulate the data and create a date time index.

beef['Year'] = pd.to_datetime(beef['Year'], format="%Y")
beef
```

```python
beef['Date'] = beef.apply(lambda x:(x['Year'] + BQuarterBegin(x['Quarter'])), axis=1)
beef
```

```python
beef.drop(['Year', 'Quarter'], axis=1, inplace=True)
beef.set_index('Date', inplace=True)
beef
```

```python
# RLS estimation
# The Recursive Least Squares (RLS) method is used for estimating time-varying coefficients 
# and is particularly useful in cases where the relationship between the variables might change over time.
# RLS is designed to handle data that changes over time. It continuously updates the model parameters as new data points are added.

# Set the dependent (endogenous) variable to the 'Quantity' column from the beef dataset
endog = beef['Quantity']

# Set the independent (exogenous) variable to the 'Price' column from the beef dataset
# Also add a constant term to the independent variable (intercept)
exog = sm.add_constant(beef['Price'])

# Define a Recursive Least Squares model using the dependent and independent variables
mod = sm.RecursiveLS(endog, exog)

# Fit the Recursive Least Squares model to the data
res = mod.fit()

# Print the summary of the model fit, which includes statistics and coefficients
print(res.summary())

# The RLS model computes the regression parameters recursively, so there are as many estimates as there are data points, 
# the summary table only presents the regression parameters estimated on the entire sample; these estimates are equivalent to OLS estimates.
```

```python
mod.k_exog # number of exogenous (independent) variables in the model. 
           # Essentially, it tells you how many predictors (including the intercept, if added) are being used in your regression model.
```

```python
# RLS plots
# We can generate the recursively estimated coefficients plot on a given variable.

res.plot_recursive_coefficient(range(mod.k_exog), alpha=None, figsize=(10,6));

# used to visualize how the model's coefficients change over time as new data points are added incrementally.
```

```python
# For convenience, we visually check for parameter stability using the plot_cusum function.

fig = res.plot_cusum(figsize=(10,6));
```

The plot shows the cumulative sum of recursive residuals, which should ideally stay within certain boundaries if the model is stable. The boundaries (often at the 5% significance level) indicate the range within which the cumulative sum should stay if the model is stable.

By plotting the CUSUM of recursive residuals, you can see if there are any periods where the relationship between price and quantity has changed significantly. If the CUSUM line stays within the boundaries, you can be confident that the model is stable. If it crosses the boundaries, it may indicate that some external factors have caused a shift in the relationship.

N.B. In the CUSUM plot, a point on the cumulative sum line crossing the zero line doesn't mean the residual at that specific point is zero. Instead, it shows the overall cumulative sum of residuals up to that point.
