```python
# dataset

import pandas as pd
import numpy as np

# Set a random seed for reproducibility
np.random.seed(0)

# Generate synthetic data for acceleration and altitude
acceleration = np.random.normal(0, 1, 1000)
altitude = np.random.normal(0, 1, 1000)

# Generate synthetic data for fuel usage
# Assume that fuel usage is a linear function of acceleration and altitude plus some noise
fuel_usage = 3*acceleration + 2*altitude + np.random.normal(0, 0.1, 1000)

# Create a pandas DataFrame
df = pd.DataFrame({
    'Acceleration': acceleration,
    'Altitude': altitude,
    'Fuel_Usage': fuel_usage
})

print(df.head())
df.shape
```

```python
X = df[['Acceleration','Altitude']]
X
```

```python
y = df['Fuel_Usage']
y
```

```python
import seaborn as sns

sns.scatterplot(x=X['Acceleration'], y=X['Altitude'], hue=y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 33)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
class LinearRegression:
    
    def __init__(self):
        # The __init__ method is the constructor for the class. It gets called when an object of this class is instantiated.
        # Here, we initialize the coefficients to None. These are the parameters of the linear regression model that we will learn during the fitting process.
        self.coefficients = None

    def fit(self, X, y):
        # The fit method is used to train the linear regression model.
        # X is the matrix of input features and y is the vector of target values.

        # We add a column of ones to the input matrix X. This is done to include an intercept term in our linear model.
        # Without this, our model would be forced to go through the origin, which is rarely a good fit to the data.
        X = np.concatenate([np.ones((X.shape[0], 1)), X], axis=1)

        # The '@' operator represents matrix multiplication in Python. It's equivalent to the dot product for 1-D arrays.
        # The expression X.T @ X computes the dot product of the transpose of X (X.T) and X itself. This is part of the normal equation for linear regression.
        # The np.linalg.inv function computes the inverse of a matrix. This is also part of the normal equation.
        # The normal equation is a mathematical formula that gives the solution to the linear regression problem with the least squares cost function.
        # It states that the coefficients that minimize the sum of squared residuals (the difference between the actual and predicted values) are given by (X^T * X)^-1 * X^T * y.
        self.coefficients = np.linalg.inv(X.T @ X) @ X.T @ y
        self.coefficients = self.coefficients.reshape(-1, 1)


    def predict(self, X):
        # The predict method is used to make predictions with the trained model.
        # X is the matrix of input features.

        # We add a column of ones to the input matrix X, just like we did in the fit method. This is because our model was trained with this additional intercept term.
        X = np.concatenate([np.ones((X.shape[0], 1)), X], axis=1)

        # We compute the predictions as the dot product of the input matrix X and the learned coefficients.
        # This is the standard formula for a linear regression prediction: y = X * beta, where beta are the coefficients.
        # The '@' operator is used for matrix multiplication, so X @ self.coefficients computes the dot product of X and the coefficients.
        # return X @ self.coefficients
        return np.dot(X, self.coefficients)

```

```python
model_lr = LinearRegression()

model_lr
```

```python
model_lr.fit(X_train, y_train)
```

```python
y_hat = model_lr.predict(X_test)

y_hat[:10]
```

```python
def compute_gradient(model, x_array, h=1e-7):
    # Initialize a zero vector for the gradient
    grad = np.zeros_like(x_array)
    
    # Iterate over each dimension of the input vector x_array
    for i in range(len(x_array)):
        # Store the original value of x_array[i]
        tmp_val = x_array[i]
        
        # Compute the function value f(x_array+h)
        x_array[i] = tmp_val + h
        fxh1 = model.predict(x_array.reshape(1, -1))  # Reshape x_array
        
        # Compute the function value f(x-h)
        x_array[i] = tmp_val - h
        fxh2 = model.predict(x_array.reshape(1, -1))  # Reshape x_array
        
        # Compute the gradient (central difference)
        grad[i] = (fxh1 - fxh2) / (2*h)
        
        # Restore the value of x[i]
        x_array[i] = tmp_val
    
    return grad
```

```python
def minimize(model, X0, X_min, X_max):
    # Initialize the independent variables with the initial guess
    min_X = X0

    # Set a small step size for the gradient descent
    step_size = 0.01

    # Repeat the optimization for a fixed number of iterations
    for _ in range(1000):
        # Compute the gradient of the function at the current point
        gradient = compute_gradient(model, np.array(min_X))

        # Update the independent variables by taking a step in the direction of steepest descent
        min_X = min_X - step_size * gradient

        # Check if the new values are within the allowed range
        min_X = np.clip(min_X, X_min, X_max)

    # Compute the minimum function value
    min_y = model.predict(np.array(min_X).reshape(1, -1))

    # Return the values of the independent variables that resulted in the minimum function value
    return min_X, min_y
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

min_value_feature1 = min(X.Acceleration)
max_value_feature1 = max(X.Acceleration)
min_value_feature2 = min(X.Altitude)
max_value_feature2 = max(X.Altitude)

min_value_feature1, max_value_feature1, '---', min_value_feature2, max_value_feature2
```

```python
# Initial guess for the independent variables
X0 = [0, 0]

X_min = [min_value_feature1, min_value_feature2]
X_max = [max_value_feature1, max_value_feature2]

# Call the minimize function
min_X, min_y = minimize(model_lr, X0, X_min, X_max)


# Print the acceleration and altitude that result in the minimum predicted fuel usage
print(f"Minimum fuel usage: {min_y}")
print(f"Acceleration: {min_X[0]}, Altitude: {min_X[1]}")
```

```python
from scipy.optimize import minimize

# Define the objective function that we want to minimize.
# This function takes in an acceleration value and returns the predicted fuel usage.
def objective(acceleration, altitude, model):
    # Create a feature vector with the given acceleration and altitude
    X = np.array([[acceleration, altitude]])
    # Use the model to predict the fuel usage
    fuel_usage = model_lr.predict(X)
    return fuel_usage

```

```python
# Now, for a given altitude, we can find the acceleration that minimizes the fuel usage.
altitude = 0.5  # replace with the desired altitude
initial_guess = 0
bounds = [(-3, 3)]  # bounds for acceleration

result = minimize(objective, initial_guess, args=(altitude, model_lr), bounds=bounds)

# The optimal acceleration is stored in the 'x' attribute of the result
optimal_acceleration = result.x

altitude, result.x, result.fun
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the data points
ax.scatter(X_test['Acceleration'], X_test['Altitude'], y_test, color='yellow', alpha=0.6, label='True fuel consumption')
ax.scatter(X_test['Acceleration'], X_test['Altitude'], y_hat, color='green', alpha=0.9, label='Predicted fuel consumption')

# Highlight the point with the optimal acceleration
ax.scatter([optimal_acceleration], [altitude], [result.fun], color='red', s=100, label='Optimal point')

ax.set_xlabel('Acceleration')
ax.set_ylabel('Altitude')
ax.set_zlabel('Fuel Usage')

plt.legend()
plt.show()
```
