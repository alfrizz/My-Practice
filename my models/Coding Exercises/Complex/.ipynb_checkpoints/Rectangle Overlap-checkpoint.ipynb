You are given two rectangles a and b each defined by four ordered pairs denoting their corners on the x, y plane. Write a function rectangle_overlap to determine whether or not they overlap. Return True if so, and False otherwise.

Note: If the two rectangles border one another, or share a corner like two diagonally adjacent positions on a chessboard, they are said to overlap.

Note: The lists of ordered pairs are in no particular order. The first entry in list a could be the top left corner while the first in list b is the bottom right.

Example:

Input:

a = [(-3,5), (-3,2),(0,5),(0,2)]

b = [(-1,4), (3,4), (3,1), (-1,1)]

Output:

def rectangle_overlap(a, b) -> True

As point (0,2) is fully contained in rectangle b and point (-1,4) is fully contained in a rectangle a.

```python
import matplotlib.pyplot as plt
```

```python
def draw_rect(x_low, x_high, y_low, y_high):

    # Calculate the corners
    points = [(x_low, y_low), (x_high, y_low), (x_high, y_high), (x_low, y_high)]

    # Unzip the list of points
    x, y = zip(*points)

    # Close the rectangle by adding the first point at the end
    x = list(x) + [x[0]]
    y = list(y) + [y[0]]
    
    # print('x',x)  ################
    # print('y',y)   ################

    plt.plot(x, y, 'bo-')  # 'bo-' means blue color, circle markers, solid line
    plt.fill(x, y, 'lightblue', alpha=0.5)  # Fill with color and set transparency
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Rectangle Plot')
    plt.grid(True)
```

```python
def retrieve_edges(rect, i, edges):
    
    x_low = min([point[0] for point in rect])
    edges[i]['x_low'] = x_low

    x_high = max([point[0] for point in rect])
    edges[i]['x_high'] = x_high

    y_low = min([point[1] for point in rect])
    edges[i]['y_low'] = y_low

    y_high = max([point[1] for point in rect])
    edges[i]['y_high'] = y_high
        
    return x_low, x_high, y_low, y_high, edges
```

```python
def rectangle_overlap(rect_a, rect_b):
    
    edges = [{}, {}] # list to save the edges x_low, x_high, y_low, y_high in two dictionaries: 
                # of rect_a (dict in pos 0) 
                # of rect_b (dict in pos 1)
    
    for i, rect in enumerate((rect_a, rect_b)):
        x_low, x_high, y_low, y_high, edges = retrieve_edges (rect, i, edges)

        draw_rect(x_low, x_high, y_low, y_high)

    plt.figure()
    plt.show()
    
    print(edges)
    
    if edges[0]['x_low'] > edges[1]['x_high'] or \
     edges[1]['x_low'] > edges[0]['x_high'] or \
     edges[0]['y_low'] > edges[1]['y_high'] or \
     edges[1]['y_low'] > edges[0]['y_high']: # if at least one of these conditions is met, there is no overlap
        
        return False
    
    else:
        
        return True
    
```

```python
a = [(-3,5), (-3,2),(0,5),(0,2)]

b = [(-1,4), (3,4), (3,1), (-1,1)]

rectangle_overlap(a, b)
```
