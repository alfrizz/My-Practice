```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Original data
x = np.random.rand(4000) * 10

plt.hist(x, 100)
```

```python
# Apply Box-Cox transformation
x_transformed, lambda_best_fit = stats.boxcox(x)

print(f"The best-fit lambda: {lambda_best_fit}")

plt.hist(x_transformed, 100)
```

```python

# In this code, the shapiro function from scipy.stats is used to perform the Shapiro-Wilk test. The function returns two values:

# W: This is the test statistic. The closer it is to 1, the more likely the data came from a normal distribution.
# p: This is the p-value. If the p-value is less than your chosen significance level (often 0.05), then you can reject the null hypothesis and conclude that the data is not normally distributed.

# In the context of the Shapiro-Wilk test, the null hypothesis is that the data is normally distributed. So, a small p-value (much less than 0.05) indicates that we have strong evidence to reject the null hypothesis. This means we conclude that the data does not come from a normal distribution.

import numpy as np
from scipy import stats


# Apply the Shapiro-Wilk test to the original data
W, p = stats.shapiro(x)

print(f"W: {W}, p-value: {p}")
```

```python
# Apply the Shapiro-Wilk test to the transformed data
W, p = stats.shapiro(x_transformed)

print(f"W: {W}, p-value: {p}")
```

```python
# In the Q-Q plots, the quantiles of your data are plotted against the quantiles of a normal distribution. If your data is normally distributed, the points in the Q-Q plot will fall along a straight line. Deviations from this straight line suggest departures from normality. 

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Q-Q plot for the original data
stats.probplot(x, plot=axs[0])
axs[0].set_title('Q-Q Plot for Original Data')

# Q-Q plot for the transformed data
stats.probplot(x_transformed, plot=axs[1])
axs[1].set_title('Q-Q Plot for Transformed Data')

plt.show()
```

```python
# lets consider a dataset that follows an exponential distribution. This distribution is not normal, but it can often be transformed into a normal distribution using the Box-Cox transformation.



# Generate data from an exponential distribution
np.random.seed(0)  # for reproducibility
x = np.random.exponential(scale=2, size=1000)

# Apply Box-Cox transformation
x_transformed, lambda_best_fit = stats.boxcox(x)

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Histogram for the original data
axs[0].hist(x, bins=50)
axs[0].set_title('Histogram for Original Data')

# Histogram for the transformed data
axs[1].hist(x_transformed, bins=50)
axs[1].set_title('Histogram for Transformed Data')

plt.show()

# Apply the Shapiro-Wilk test to the original data
W, p = stats.shapiro(x)
print(f"Original data: W = {W}, p-value = {p}")

# Apply the Shapiro-Wilk test to the transformed data
W, p = stats.shapiro(x_transformed)
print(f"Transformed data: W = {W}, p-value = {p}")
```

```python
# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# Q-Q plot for the original data
stats.probplot(x, plot=axs[0])
axs[0].set_title('Q-Q Plot for Original Data')

# Q-Q plot for the transformed data
stats.probplot(x_transformed, plot=axs[1])
axs[1].set_title('Q-Q Plot for Transformed Data')

plt.show()
```
