```python
import sklearn
from sklearn.datasets import load_diabetes
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
diabetics = load_diabetes()
column_names = diabetics.feature_names
df_diabetics = pd.DataFrame(diabetics.data, columns=column_names)
print(df_diabetics.head())

# Visualize the distribution of a specific feature (e.g., 'age')
plt.figure(figsize=(8, 6))
plt.title("Distribution original")
plt.hist(df_diabetics.iloc[:, 0], bins=20, color='brown', alpha=0.7)
plt.xlabel("Standardized 'age'")
plt.ylabel("Frequency")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
X_standardized = scaler.fit_transform(df_diabetics)

# Visualize the distribution of a specific feature (e.g., 'age')
plt.figure(figsize=(8, 6))
plt.title("Distribution after Standardization")
plt.hist(X_standardized[:, 0], bins=20, color='blue', alpha=0.7)
plt.xlabel("Standardized 'age'")
plt.ylabel("Frequency")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Normalize the data
minmax_scaler = MinMaxScaler()
X_normalized = minmax_scaler.fit_transform(df_diabetics)

# Visualize the distribution of the same feature after normalization
plt.figure(figsize=(8, 6))
plt.title("Distribution after Normalization")
plt.hist(X_normalized[:, 0], bins=20, color='green', alpha=0.7)
plt.xlabel("Normalized 'age'")
plt.ylabel("Frequency")
plt.show()
```
