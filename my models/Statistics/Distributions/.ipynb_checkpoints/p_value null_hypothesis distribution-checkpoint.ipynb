```python
import numpy as np
from scipy.stats import binom_test
import matplotlib.pyplot as plt

# Number of tosses in each experiment
n_tosses = 1000

# Number of experiments
n_experiments = 10000

# List to store p-values
p_values = []

# Perform the experiments
for _ in range(n_experiments):
    # Toss the coin n_tosses times (1=head, 0=tail)
    tosses = np.random.choice([0, 1], size=n_tosses)

    # Count the number of heads
    n_heads = np.sum(tosses)

    # Perform a binomial test and get the p-value
    # The binomial test checks if the number of successes (heads) is statistically different from what would be expected under the null hypothesis (a fair coin)
    p_value = binom_test(n_heads, n_tosses, p=0.5)
    
    # Add the p-value to the list
    p_values.append(p_value)

# Plot a histogram of the p-values
plt.hist(p_values, bins=20, alpha=0.75)
plt.xlabel('p-value')
plt.ylabel('Frequency')
plt.title('Histogram of p-values from multiple experiments')
plt.grid(True)
plt.show()
```
