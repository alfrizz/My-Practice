```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import geom

# You have a group of couples that decide to have children until they have their first girl, after which they stop having children. What is the expected gender ratio of the children that are born? What is the expected number of children each couple will have?

def plot_geom_pmf(p, num_trials):
    
    exp_val = 0 # expected mean value of the distribution: n_children
    exp_ratio = 0 # expected mean ratio of possible outcomes: n_boys / n_girls = (n_children - 1) / 1
    
    # Define the range of x values
    x = np.arange(1, num_trials+1)

    # Calculate PMF
    pmf = geom.pmf(x, p)

    # Create the plot
    plt.figure(figsize=(10,6))
    plt.bar(x, pmf)

    # Add y value over each bar
    for i in range(num_trials):
        exp_val += x[i]*pmf[i]
        exp_ratio += ((x[i]-1)/1*pmf[i]) # x[i] boys / 1 girl
        plt.text(x[i], pmf[i], round(pmf[i], 2), ha = 'center')

    plt.title('PMF of Geometric Distribution')
    plt.xlabel('Number of Trials')
    plt.ylabel('Probability')
    plt.show()
    
    print ('Expected Value (average children):', exp_val)
    print ('Expected Ratio (boys/girls):', exp_ratio)

# Call the function with your desired probability and number of trials
plot_geom_pmf(0.5, 12)
```
