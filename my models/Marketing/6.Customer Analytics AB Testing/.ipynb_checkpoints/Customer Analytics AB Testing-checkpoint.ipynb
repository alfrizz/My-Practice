Exploring the power calculation

As discussed, power is the probability of rejecting the null hypothesis when the alternative hypothesis is true. Here you will explore some properties of the power function and see how it relates to sample size among other parameters. 

The get_power() function has been included and takes the following arguments in the listed order: n for sample size, p1 as the baseline value, p2 as the value with lift included, and cl as the confidence level.

```python

def get_power(sample_size, control_prop, treatment_prop, conf_level):
    """
    Calculate the statistical power of a two-proportion z-test.

    Parameters:
    sample_size (int): Total number of samples.
    control_prop (float): Proportion in the control group.
    treatment_prop (float): Proportion in the treatment group.
    conf_level (float): Confidence level (e.g., 0.95 for 95% confidence).

    Returns:
    float: The power of the test.
    """
    
    # Calculate alpha (significance level)
    alpha = 1 - conf_level
    
    # Calculate the critical value for the normal distribution (z-score)
    z_critical = stats.norm.ppf(1 - alpha / 2)
    
    # Calculate the absolute difference between the two proportions
    prop_diff = abs(treatment_prop - control_prop)
    
    # Calculate the pooled proportion
    pooled_prop = (control_prop + treatment_prop) / 2
    
    # Calculate the variances for each proportion
    var_control = control_prop * (1 - control_prop)
    var_treatment = treatment_prop * (1 - treatment_prop)
    
    # Calculate the standard error for the difference between proportions
    standard_error = ((var_control + var_treatment) / sample_size) ** 0.5
    
    # Calculate the z-score for the effect size
    z_effect_size = prop_diff / standard_error
    
    # Calculate beta components using the cumulative distribution function (CDF)
    # The formula below captures the Type II error by considering both tails of the distribution under the alternative hypothesis:
    beta = stats.norm.cdf(z_critical - z_effect_size) - stats.norm.cdf(-z_critical - z_effect_size)
    
    # Calculate the power of the test
    power = 1 - beta
    
    return power
```

```python
# same function with statsmodel libraries, to verify the results

def get_power_statsmodels(sample_size, control_prop, treatment_prop, conf_level):
    """
    Calculate the statistical power of a two-proportion z-test using statsmodels.

    Parameters:
    sample_size (int): Total number of samples.
    control_prop (float): Proportion in the control group.
    treatment_prop (float): Proportion in the treatment group.
    conf_level (float): Confidence level (e.g., 0.95 for 95% confidence).

    Returns:
    float: The power of the test.
    """
    # Calculate alpha (significance level)
    alpha = 1 - conf_level

    # Calculate the effect size based on proportions
    effect_size = sms.proportion_effectsize(control_prop, treatment_prop)

    # Calculate power using statsmodels
    power_analysis = sms.NormalIndPower()
    power = power_analysis.power(effect_size=effect_size, nobs1=sample_size, alpha=alpha, ratio=1)

    return power
```

```python
# Example parameters
control_prop = 0.1
treatment_prop = 0.12

# Test impact of sample size increase on power
sample_size_1 = 1000
power_1 = get_power(sample_size_1, control_prop, treatment_prop, 0.95)
power_1_statsmodels = get_power_statsmodels(sample_size_1, control_prop, treatment_prop, 0.95)
print(f"Power with sample size {sample_size_1}: {power_1} (Custom), {power_1_statsmodels} (Statsmodels)")

sample_size_2 = 2000
power_2 = get_power(sample_size_2, control_prop, treatment_prop, 0.95)
power_2_statsmodels = get_power_statsmodels(sample_size_2, control_prop, treatment_prop, 0.95)
print(f"Power with sample size {sample_size_2}: {power_2} (Custom), {power_2_statsmodels} (Statsmodels)")

# Test impact of confidence level increase on power
conf_level_1 = 0.80
power_conf_1 = get_power(sample_size_1, control_prop, treatment_prop, conf_level_1)
power_conf_1_statsmodels = get_power_statsmodels(sample_size_1, control_prop, treatment_prop, conf_level_1)
print(f"Power with confidence level {conf_level_1}: {power_conf_1} (Custom), {power_conf_1_statsmodels} (Statsmodels)")

conf_level_2 = 0.95
power_conf_2 = get_power(sample_size_1, control_prop, treatment_prop, conf_level_2)
power_conf_2_statsmodels = get_power_statsmodels(sample_size_1, control_prop, treatment_prop, conf_level_2)
print(f"Power with confidence level {conf_level_2}: {power_conf_2} (Custom), {power_conf_2_statsmodels} (Statsmodels)")

# Compare the ratios
print(f"Ratio of powers for sample sizes: {power_2 / power_1}")
print(f"Ratio of powers for confidence levels: {power_conf_1 / power_conf_2}")
```

```python
def get_sample_size(power, p1, p2, cl, max_n=1000000):
    n = 1 
    while n <= max_n:
        tmp_power = get_power(n, p1, p2, cl)

        if tmp_power >= power: 
            return n 
        else: 
            n = n + 100

    return "Increase Max N Value"
```

Intuition behind statistical significance

In this exercise you will work to gain an intuitive understanding of statistical significance. You will do this by utilizing the get_pvalue() function on a variety of parameter sets that could reasonably arise or be chosen during the course of an A/B test. While doing this you should observing how statistical significance results vary as you change the parameters. This will help build your intuition surrounding this concept, and reveal some of the subtle pitfalls of p-values.

```python
def get_pvalue(con_conv, test_conv, con_size, test_size):
    """
    Calculate the p-value for the difference in conversion rates between control and test groups using a z-test.

    Parameters:
    con_conv (float): Conversion rate in the control group.
    test_conv (float): Conversion rate in the test group.
    con_size (int): Number of observations in the control group.
    test_size (int): Number of observations in the test group.

    Returns:
    float: The p-value for the difference in conversion rates.
    """
    
    # Calculate the pooled proportion
    pooled_prop = (con_conv * con_size + test_conv * test_size) / (con_size + test_size)
    
    # Calculate the standard error of the difference in proportions
    standard_error = ((pooled_prop * (1 - pooled_prop) * (1 / con_size + 1 / test_size)) ** 0.5)
    
    # Calculate the z-statistic
    z_statistic = (test_conv - con_conv) / standard_error
    
    # Calculate the two-tailed p-value
    p_value = 2 * stats.norm.sf(abs(z_statistic))
    
    return p_value
```

```python
p_value = get_pvalue(con_conv=0.1, test_conv=0.17, con_size=1000, test_size=1000)
print(p_value)
```

```python
# Get the p-value
p_value = get_pvalue(con_conv=0.1, test_conv=0.15, con_size=100, test_size=100)
print(p_value)
```

```python
# Get the p-value
p_value = get_pvalue(con_conv=0.48, test_conv=0.50, con_size=1000, test_size=1000)
print(p_value)
```

In this exercise, you'll develop your intuition for how various parameter values impact confidence intervals. Specifically, you will explore through the get_ci() function how changes widen or tighten the confidence interval. This is the function signature, where cl is the confidence level and sd is the standard deviation.

```python
def get_ci(value, cl, sd, sample_size):
    """
    Calculate the confidence interval for a given mean, confidence level, standard deviation, and sample size.

    Parameters:
    value (float): The sample mean.
    cl (float): Confidence level (e.g., 0.95 for 95% confidence).
    sd (float): Standard deviation of the sample.
    sample_size (int): Size of the sample.

    Returns:
    tuple: The lower and upper bounds of the confidence interval.
    """
    
    # Calculate alpha (significance level)
    alpha = 1 - cl
    
    # Calculate the z-score for the given confidence level
    z_score = stats.norm.ppf(1 - alpha / 2)
    
    # Calculate the standard error
    se = sd / (sample_size ** 0.5)
    
    # Calculate the margin of error
    moe = z_score * se
    
    # Calculate the lower and upper bounds of the confidence interval
    lwr_bnd = value - moe
    upr_bnd = value + moe
    
    return (lwr_bnd, upr_bnd)
```

```python
# Compute and print the confidence interval
confidence_interval  = get_ci(1, 0.975, 0.5, 2000)
print(confidence_interval)
```

```python
# Compute and print the confidence interval
confidence_interval  = get_ci(1, 0.90, 0.5, 2000)
print(confidence_interval)
```

```python
# Compute and print the confidence interval
confidence_interval  = get_ci(1, 0.975, 2, 2000)
print(confidence_interval)
```

```python
# Compute and print the confidence interval
confidence_interval  = get_ci(1, 0.95, 0.001, 2000)
print(confidence_interval)
```
