```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import poisson
```

Fitness goals

Let's model how activity levels impact weight loss using modern fitness trackers. On days when you go to the gym, you average around 15k steps, and around 5k steps otherwise. You go to the gym 40% of the time. Let's model the step counts using different distributions

```python
# 1) Represent the steps using a mixture of two distributions with these parameters

# Parameters
mu_gym = 15000
sigma_gym = 1000  # assumed

mu_nongym = 5000
sigma_nongym = 500  # assumed

p_gym = 0.4
p_nongym = 0.6

# Generate distributions
n_days = 1000
```

```python
# 1) Represent the steps using a mixture of two normal distributions

gym_days_norm = np.random.normal(mu_gym, sigma_gym, int(n_days * p_gym))
nongym_days_norm = np.random.normal(mu_nongym, sigma_nongym, int(n_days * p_nongym))

# Plot distributions
plt.hist(gym_days_norm, bins=30, alpha=0.5, color='blue', label='Gym Days')
plt.hist(nongym_days_norm, bins=30, alpha=0.5, color='green', label='Non-Gym Days')

# Add titles and labels
plt.title('Normal Distributions of Daily Steps')
plt.xlabel('Steps')
plt.ylabel('Frequency')
plt.legend()
```

```python
# Combine distributions
steps = np.concatenate([gym_days_norm, nongym_days_norm])


# Plot
plt.hist(steps, bins=90, alpha=0.7, color='blue')
plt.title(' Normal Distribution of Daily Steps')
plt.xlabel('Steps')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2) Represent the steps using a mixture of two Poisson distribution

gym_days_poiss = np.random.poisson(mu_gym, int(n_days * p_gym))
nongym_days_poiss = np.random.poisson(mu_nongym, int(n_days * p_nongym))


# Plot distributions
plt.hist(gym_days_poiss, bins=30, alpha=0.5, color='blue', label='Gym Days', density=True)
plt.hist(nongym_days_poiss, bins=30, alpha=0.5, color='green', label='Non-Gym Days', density=True)

# Add titles and labels
plt.title('Poisson Distribution of Daily Steps')
plt.xlabel('Steps')
plt.ylabel('Frequency')
plt.legend()
```

```python
# 3) Represent the steps using a mixture of two LogNormal distribution

import numpy as np
import matplotlib.pyplot as plt

# Lognormal parameters
mu_gym_log = np.log(mu_gym**2 / np.sqrt(sigma_gym**2 + mu_gym**2))
sigma_gym_log = np.sqrt(np.log(1 + (sigma_gym**2 / mu_gym**2)))

mu_nongym_log = np.log(mu_nongym**2 / np.sqrt(sigma_nongym**2 + mu_nongym**2))
sigma_nongym_log = np.sqrt(np.log(1 + (sigma_nongym**2 / mu_nongym**2)))



print(mu_gym_log, sigma_gym_log)
print(mu_nongym_log, sigma_nongym_log)

# eventually increase the standard deviations to make the distributions more skewed
```

```python
gym_days_lognorm = np.random.lognormal(mu_gym_log, sigma_gym_log, int(n_days * p_gym))
nongym_days_lognorm = np.random.lognormal(mu_nongym_log, sigma_nongym_log, int(n_days * p_nongym))

# Plot distributions
plt.hist(gym_days_lognorm, bins=30, alpha=0.5, color='blue', label='Gym Days')
plt.hist(nongym_days_lognorm, bins=30, alpha=0.5, color='green', label='Non-Gym Days')

# Add titles and labels
plt.title('Lognormal Distributions of Daily Steps')
plt.xlabel('Steps')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
steps_log = np.concatenate([gym_days_lognorm, nongym_days_lognorm])
steps_log.shape
```

For simplicity, lets say you have an 80% chance of losing 1lb and a 20% chance of gaining 1lb when you get more than 10k steps. The probabilities are reversed when you get less than 8k steps. Otherwise, there's an even chance of gaining or losing 1lb. Given all this information, find the probability of losing weight in a month.

```python
days_losing_weight = []
n_days = steps_log.shape[0]

for steps_per_day in range(n_days):
    if steps_per_day > 10000:
        p = [0.8,0.2]
    elif steps_per_day < 8000:
        p = [0.2,0.8]
    else:
        p = [0.5,0.5]
        
    days_losing_weight.append(np.random.choice([1,0], p=p))
                              
prob_losing_weight = sum(days_losing_weight) / n_days
prob_losing_weight
```
