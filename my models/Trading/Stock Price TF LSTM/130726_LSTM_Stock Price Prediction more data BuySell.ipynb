```python
def create_arima_features (ARIMA_fields, time_steps, df_input, SMA_window, AR=True, SMA=True, ID1 = True, ID2 = True):
    

    ARIMA_features = []
    
    for field in ARIMA_fields:
            
        if AR: # calculate Auto Regressive fields          
            AR_field_name = field+' AR_'+str(time_steps)+'TS'
            # df_input[AR_field_name] = [np.round(df_input[field+' Scaled'][i-time_steps:i].values, 6) 
            df_input[AR_field_name] =  [np.round(df_input[field+' Scaled'].shift(1)
                                        .iloc[i-time_steps+1:i+1].values, 6)
                                        for i in range(len(df_input))] 
            ARIMA_features.append(AR_field_name)
   
        if SMA: # calculate Simple Moving Average fields (for a user defined window)
            SMA_field_name= field+' SMA_'+str(time_steps)+'TS_'+str(SMA_window)+'WD'
            df_input[SMA_field_name] = [np.round(df_input[field+' Scaled'].shift(1).rolling(window=SMA_window).mean()
                                        .iloc[i-time_steps+1:i+1].values, 6)
                                        for i in range(len(df_input))]
            ARIMA_features.append(SMA_field_name)
            
            
        if ID1: # calculate the first Integrated Differencing fields 
            ID1_field_name = field+' ID1_'+str(time_steps)+'TS'
            df_input[ID1_field_name] = [np.round((df_input[field+' Scaled'].shift(1) - df_input[field+' Scaled'].shift(2))
                                       .iloc[i-time_steps+1:i+1].values, 6)
                                        for i in range(len(df_input))]
            ARIMA_features.append(ID1_field_name)
            ID1_simplefield_name = field+' ID1_simple' # collumns used later to calculate ID2
            df_input[ID1_simplefield_name] = np.round((df_input[field+' Scaled'].shift(1) - df_input[field+' Scaled'].shift(2)) ,6)
            # ARIMA_features.append(ID1_simplefield_name)
    
    if (ID1 and ID2): # calculate the second Integrated Differencing fields
        for field in df_input.filter(like='ID1_simple').columns:
            ID2_field_name = field.replace('ID1_simple', '')+' ID2_'+str(time_steps)+'TS'
            df_input[ID2_field_name] = [np.round((df_input[field] - df_input[field].shift(1))
                                           .iloc[i-time_steps+1:i+1].values, 6)
                                            for i in range(len(df_input))]
            ARIMA_features.append(ID2_field_name)
    
    # remove the first rows and the ID1 simple columns
    cols_to_drop = df_input.filter(like='ID1_simple').columns
    rows_to_filt = time_steps+SMA_window
    df_output = df_input.drop(columns=cols_to_drop, axis=1)
    return df_output.iloc[rows_to_filt:,:], ARIMA_features
```

```python
ind_split = int(np.round(len(df) * (1-test_size)))

# select the fields to use, and create another dataframe containing only these fields

df_feat = df[ARIMA_fields]
df_feat_train = df_feat.iloc[:ind_split,:]
df_feat_test = df_feat.iloc[ind_split:,:]

# fit the scaler on df_feat_train and transform on df_feat_test (for each feature, a different scaler), then save all scalers
scalers = {}
df_feat_train_scal = pd.DataFrame(index = df_feat_train.index)
df_feat_test_scal = pd.DataFrame(index = df_feat_test.index)

for field in ARIMA_fields:
    scaler = StandardScaler()
    df_feat_train_scal[field+' Scaled'] = scaler.fit_transform(np.float64(df_feat_train[[field]]))
    df_feat_test_scal[field+' Scaled'] = scaler.transform(df_feat_test[[field]])
    scalers[field] = scaler

# merge df_feat_train_scal and df_feat_test_scal,
df_feat_full_scal = pd.concat([df_feat_train_scal, df_feat_test_scal], axis = 0)

# add the 'Date' and 'Adjusted Close' columns to the df_feat_full_scal, that will be used later for visualization
df_feat_full_scal['Date'] = df['Date'].tolist()
df_feat_full_scal['Adjusted Close'] = df['Adjusted Close'].tolist()

df_ARIMA, ARIMA_features = create_arima_features (ARIMA_fields = ARIMA_fields, 
                                                  time_steps = time_steps, 
                                                  df_input = df_feat_full_scal,
                                                  SMA_window = SMA_window,
                                                  AR = True,
                                                  SMA = True,
                                                  ID1 = ID1,
                                                  ID2 = ID2)
    
# create features dataframes
X_train = df_ARIMA[ARIMA_features].iloc[:ind_split,:]
X_test = df_ARIMA[ARIMA_features].iloc[ind_split:,:]     
    
# create labels dataframes
y_train = df_ARIMA['Adjusted Close Scaled'].iloc[:ind_split]
y_test = df_ARIMA['Adjusted Close Scaled'].iloc[ind_split:]  

# dataframes used later for visualization
df_train = df_ARIMA.iloc[:ind_split]
df_test = df_ARIMA.iloc[ind_split:]  
    
X_train.shape, X_test.shape, y_train.shape, y_test.shape,


# X_train.head()
# X_test.tail()
```

```python
X_train_list = []  
for n_feat in range(X_train.shape[1]):
    X_train_list.append(np.array(X_train.iloc[:,n_feat].values.tolist()))
X_train_resh = np.dstack(X_train_list)

X_test_list = []
for n_feat in range(X_test.shape[1]):
    X_test_list.append(np.array(X_test.iloc[:,n_feat].values.tolist()))
X_test_resh = np.dstack(X_test_list)

y_train_resh = np.array(y_train).reshape(-1 , 1)

y_test_resh = np.array(y_test).reshape(-1 , 1)

X_train_resh.shape , y_train_resh.shape, X_test_resh.shape, y_test_resh.shape
```

```python
model = keras.models.Sequential()


model.add(keras.layers.Bidirectional(keras.layers.LSTM(units = 1024,
                                                       return_sequences=True,
                                                       input_shape = (X_train_resh.shape[1], 
                                                                      X_train_resh.shape[2]))))


model.add(keras.layers.Bidirectional(keras.layers.LSTM(units=512, 
                                                        return_sequences=True)))
model.add(keras.layers.Flatten()) 

model.add(keras.layers.Dropout(dropout_value))

model.add(keras.layers.Dense(1024))          
          
model.add(keras.layers.Dense(1))

# model.build(input_shape=(X_train_resh.shape[1], X_train_resh.shape[2]))
model.build(input_shape=(None,
                         X_train_resh.shape[1], 
                         X_train_resh.shape[2]))
model.summary()
```

```python
model = model

early_stopping_monitor = EarlyStopping(monitor='loss', patience=20)

optimizer = Adam(lr=learning_rate)

model.compile(optimizer=optimizer,
               # metrics=[MeanSquaredError(), MeanAbsoluteError(), RootMeanSquaredError()],
               metrics=['mse', 'mae'],
               # loss='mean_squared_error'
               loss='mse')
history = model.fit(X_train_resh,  
                    y_train_resh,
                    epochs=200,
                    batch_size=100,
                    validation_data=(X_test_resh, 
                                     y_test_resh),
                    callbacks=[early_stopping_monitor])
```

```python

def Calc_Return(init_invest, max_desired_invest = np.inf,  risk_factor = 1, commiss_perc = 0.01):
    
    Day_Amount_Invest = []
    Day_Amount_Return = []    
    Day_Short_Commiss = []
    Day_End_Amount = []
        
    for i in range(len(df_predictions)):
        
        # percentage invested proportional to the confidence of the model prediction
        perc_invest = df_predictions['Pred_variat_perc'].iloc[i] 
        #percentage of gain or loss 
        perc_gain_loss = df_predictions['Act_variat_perc'].iloc[i] 
        
        prev_amount_owned = init_invest if i == 0 else Day_End_Amount[i - 1]

        max_amount_invest = min(max_desired_invest, prev_amount_owned)
        
        amount_invest = min(prev_amount_owned * np.abs(perc_invest)/100 * risk_factor, max_amount_invest)
        # could be eventually modified, if we want to allow to invest more than the amount owned, during short trading
        Day_Amount_Invest.append(amount_invest)
        
        if commiss_perc <= 0: # no short positions allowed
            daily_return = amount_invest * perc_gain_loss/100 if np.sign(perc_invest) == 1 else 0
        else: #short positions allowed
            daily_return = amount_invest * np.abs(perc_gain_loss)/100 * np.sign(perc_invest) * np.sign(perc_gain_loss)
            # positive if sign(perc_invest)==sign(perc_gain_loss), negative otherwise
        
        Day_Amount_Return.append(daily_return)
        
        short_commiss = amount_invest * commiss_perc if np.sign(perc_invest) == -1 else 0
        Day_Short_Commiss.append(short_commiss)
        
        Day_End_Amount.append(prev_amount_owned + daily_return - short_commiss)

    return Day_Amount_Invest, Day_Short_Commiss, Day_Amount_Return, Day_End_Amount

```
