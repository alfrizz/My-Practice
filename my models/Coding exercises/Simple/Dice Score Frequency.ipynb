```python
import random
import matplotlib.pyplot as plt

def dices_score_freq(n_dices, n_throws):
    throws_score = {i:0 for i in range(n_dices, n_dices*6 + 1)}
    for j in range(n_throws):
        throw_tot = 0
        for i in range(n_dices):
            throw_tot += random.randint(1, 6)
        throws_score[throw_tot]+=1
    # plt.hist(throws_score.values(), n_dices*6)
    plt.bar(throws_score.keys(), throws_score.values())
    plt.xlabel('Scores')
    plt.ylabel('Frequency')
    plt.title('Frequency of Dice Scores')
    return throws_score

dices_score_freq(4, 1000000)
```
