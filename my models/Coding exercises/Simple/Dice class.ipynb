```python
#Dice Class - 101computing.net/oop-programming
import random

class Dice():
    #Constructor for the Dice class
    def __init__(self, numberOfSides=6):
        self.__numberOfSides = numberOfSides
        self.value = 1

    def throw(self):
        self.value = random.randint(1,self.__numberOfSides)
        return self.value
    
# #Main Program Starts here 
# dice1 = Dice(6)
# dice1.throw()
# print("First Dice:")
# print(dice1.value)

# dice2 = Dice(6)
# dice2.throw()
# print("Second Dice:")
# print(dice2.value)

# if dice1.value>dice2.value:
#     print("First dice wins!")
# elif dice1.value<dice2.value:
#     print("Second dice wins!")
# else:
#     print("It's a draw!")
```

```python
n_sides = 2
n_cards = 52
points_player1 = 0
points_player2 = 0
while n_cards > 0:

    game_over = False
    draw_space = ''
    while game_over == False:
        
        print('^^^^ cards left =', n_cards, '^^^^')
        
        dice1 = Dice(numberOfSides=n_sides)
        dice1.throw()
        print(draw_space+"First Dice:")
        print(draw_space, dice1.value)
        n_cards -= 1
        
        dice2 = Dice(numberOfSides=n_sides)
        dice2.throw()
        print(draw_space+"Second Dice:")
        print(draw_space, dice2.value)
        n_cards -= 1
        
        if dice1.value>dice2.value:
            print(draw_space+"First dice wins!")
            points_player1 += 1
            game_over = True
        elif dice1.value<dice2.value:
            print(draw_space+"Second dice wins!")
            points_player2 += 1
            game_over = True
        else:
            print(draw_space+"It's a draw!")
            draw_space+='     '
            
    print('\n*** points player 1 =', points_player1, '***')
    print('*** points player 2 =', points_player2, '***\n')
            
if points_player1 > points_player2:
    print('\n\n\nPlayer 1 WINS')
elif points_player2 > points_player1:
    print('\n\n\nPlayer 2 WINS')
else:
    print ('\n\n\nIt is a DRAW')
```
