```python
def is_palindrome (stringa):
    left_index = 0
    right_index = len(stringa)-1
    palindrome = True
    while left_index < right_index:
        if stringa[left_index]!=stringa[right_index]:
            palindrome = False
            break
        left_index +=1
        right_index -=1
    return palindrome
    
```

```python
def factorial(number):
    factorial = 1
    for fact in range(1, number+1, 1):
        factorial *= fact
    return factorial
        
        
```

```python
def factorial_rec(number):
    if number == 1:
        return 1
    return number * factorial_rec(number - 1)
    
```

```python
factorial(1500)
```

```python
factorial_rec(1500)
```

```python
def largest_element(lista: list):
    max_el = lista[0]
    for i in range(1, len(lista)):
        if lista[i] > max_el:
            max_el = lista[i]
    return max_el
```

```python
lista = [8, 5, 9, 2, 47, 6, 7, 5, 7, 42, 13, 23, 34, 22, 9, 5]
import numpy as np
largest_element(lista)
```

```python
def freq_in_list(lista: list):
    dict_freq = {}
    for elem in lista:
        if elem in dict_freq:
            dict_freq[elem] += 1
        else:
            dict_freq[elem] = 1
    return dict(sorted(dict_freq.items(), key=lambda item: item[1]))
```

```python
freq_in_list(lista)
```

```python
def is_prime(number: int):
    for i in range(2, number):
        print(number, i, number%i)
        if number%i == 0:
            return False
    return True
        
```

```python
def common_elems(lista1: list, lista2: list):
    lista1 = list(set(lista1))
    lista2 = list(set(lista2))
    common_list = []
    for elem in lista1:
        if elem in lista2:
            common_list.append(elem)
    return sorted(lista1), sorted(lista2), sorted(common_list)
```

```python
lista1 = [8, 5, 9, 2, 47, 6, 7, 5, 7, 42, 13, 23, 34, 22, 9, 5]
lista2 = [81, 51, 9, 21, 47, 61, 71, 5, 71, 62, 13, 93, 34, 22, 9, 51]

common_elems(lista1, lista2)
```

```python
def bubble_sort(lista: list):
    print(lista)
    for i in range(len(lista)):
        for j in range(len(lista) - i -1):
            if lista[j+1] < lista [j]:
                lista[j], lista [j+1] = lista[j+1], lista[j]
    return lista

    
```
