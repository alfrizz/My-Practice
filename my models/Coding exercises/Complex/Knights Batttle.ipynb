```python
import random

class Knight:
    def __init__(self, name='', health= 100):
        
        self.state = "Idle"
        self.health = health
        
        if  name == '':
            firstnames = ["Lancelot","Charles","Henry","William","James","Richard","Edward"]
            nicknames = ["Brave","Horrific","Courageous","Terrific","Fair","Conqueror","Victorious","Glorious","Invicible"]
            firstname = random.choice(firstnames)
            nickname =  random.choice(nicknames)
            self.name = firstname + " The " + nickname 
        else:
            self.name = name
    
    def pick_action(self):
        actions = ["Attack", "Defend", "Idle"]
        self.state = random.choice(actions)
    
    def draw_sprite(self):
        sprite = {
            "Idle": "Knight "+self.name+ " is standing still. Its health is "+str(self.health)+"",
            "Attack": "Knight "+self.name+" is attacking. Its health is "+str(self.health)+"",
            "Defend": "Knight "+self.name+" is defending. Its health is "+str(self.health)+""
        }
        print(sprite[self.state])
        
        
```

```python

def draw_knights(attack_comp=False, attack_opp=False, defend_comp=False, defend_opp=False):
    
    if attack_comp==False and attack_opp==False and defend_comp==False and defend_opp==False: #both idle
        knight =  [ ' O        O ',
                    '-+-      -+-',
                    '/ \\      / \\']
        
    elif attack_comp==True and attack_opp==False and defend_comp==False and defend_opp==False: #comp att, opp idle
        knight =  [ '     O  / O ',
                    '    -+-/ -+-',
                    '    / \\  / \\']
        
    elif attack_comp==True and attack_opp==True and defend_comp==False and defend_opp==False: #comp att, opp att
        knight =  [ '  O  /\\  O ',
                    ' -+-/  \\-+-',
                    ' / \\    / \\']

    elif attack_comp==True and attack_opp==False and defend_comp==False and defend_opp==True: #comp att, opp def
        knight =  [ '    O  /| O ',
                    '   -+-/ |-+-',
                    '   / \\  |/ \\']
        
    elif attack_comp==False and attack_opp==True and defend_comp==False and defend_opp==False: #comp idle, opp att
        knight =  [ ' O \\  O ',
                    '-+- \\-+-',
                    '/ \\  / \\']

    elif attack_comp==False and attack_opp==False and defend_comp==True and defend_opp==True: #comp def, opp def
        knight =  [ ' O  |   | O ',
                    '-+- |   |-+-',
                    '/ \\ |   |/ \\']

    elif attack_comp==False and attack_opp==False and defend_comp==True and defend_opp==False: #comp def, opp idle
        knight =  [ ' O  |     O ',
                    '-+- |    -+-',
                    '/ \\ |    / \\']

    elif attack_comp==False and attack_opp==False and defend_comp==False and defend_opp==True: #comp idle, opp def
        knight =  [ ' O      | O ',
                    '-+-     |-+-',
                    '/ \\     |/ \\']

    elif attack_comp==False and attack_opp==True and defend_comp==True and defend_opp==False: #comp def, opp idle
        knight =  [ ' O  |\\  O ',
                    '-+- | \\-+-',
                    '/ \\ |  / \\']

    for i in range(len(knight)):
        print(''+str(knight[i])) 


```

```python
import time
from IPython.display import clear_output

def fighting():
    state = 'starting_game' 
    
    while state == 'starting_game':    
        knight_selection = input('******* The Knights Battle *******\nSelect your knight: \nC: Custom Knight \nR: Random Knight\n')
        state = 'selecting_knight'
        while state == 'selecting_knight': 
            
            if knight_selection != 'R' and knight_selection != 'C':
                    print('Select C or R')
                    
            elif knight_selection == 'R':
                knight_companion = Knight(name='') 
                state = 'end_selecting_knight'
                
            else: # custom selection
                firstname_selection = input('Write your knight first name:\n')
                nickname_selection = input('Write your knight nickname:\n')
                full_name = firstname_selection + " The " + nickname_selection 
                state = 'selecting_health'
                while state == 'selecting_health':
                    try:
                        health_selection = input('Write your knight health (default 100):\n')
                        knight_companion = Knight(name=full_name, health=int(health_selection))
                        state = 'end_selecting_knight'
                    except ValueError: # if inputs are not numeric
                        print('Insert numeric coordinates')
                

        knight_opponent = Knight()
        print('----------------------------')
        draw_knights()
        knight_companion.draw_sprite()
        knight_opponent.draw_sprite()
        print('----------------------------')
        print('*** Starting the Battle! ***')
        print('----------------------------')
        
        while(knight_companion.health > 0 and knight_opponent.health > 0):
            knight_companion.pick_action()
            knight_opponent.pick_action()

                
            if knight_companion.state == 'Attack' and knight_opponent.state == 'Idle':
                knight_opponent.health -= 5
                draw_knights(attack_comp=True)
                
            elif knight_companion.state == 'Attack' and knight_opponent.state == 'Attack':
                knight_companion.health -= 1 
                knight_opponent.health -= 1
                draw_knights(attack_comp=True, attack_opp=True)
                
            elif knight_companion.state == 'Attack' and knight_opponent.state == 'Defend':
                knight_opponent.health -= 2
                draw_knights(attack_comp=True, defend_opp=True)
                
            elif knight_companion.state == 'Defend' and knight_opponent.state == 'Attack':
                knight_companion.health -= 2
                draw_knights(defend_comp=True, attack_opp=True)
    
            elif knight_companion.state == 'Defend' and knight_opponent.state == 'Defend':
                draw_knights(defend_comp=True, defend_opp=True)
                
            elif knight_companion.state == 'Defend' and knight_opponent.state == 'Idle':
                draw_knights(defend_comp=True)
                
            elif knight_companion.state == 'Idle' and knight_opponent.state == 'Attack':
                knight_companion.health -= 5
                draw_knights(attack_opp=True)
                
            elif knight_companion.state == 'Idle' and knight_opponent.state == 'Idle':
                draw_knights()

            elif knight_companion.state == 'Idle' and knight_opponent.state == 'Defend':
                draw_knights(defend_opp=True)

            time.sleep(0.1)
            clear_output(wait=True)
            knight_companion.draw_sprite()
            knight_opponent.draw_sprite()
            print('----------------------------')
        
        if knight_companion.health > knight_opponent.health:
            print('******************************************')
            print('*** '+str(knight_companion.name)+' WON ***')
            print('******************************************')
        else:
            print('******************************************')
            print('*** '+str(knight_companion.name)+' LOST ***')
            print('******************************************')
        
        state = 'starting_game' # to restart
```

```python
fighting()
```
