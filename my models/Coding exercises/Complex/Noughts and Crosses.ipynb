```python

def displayBanner():
    print("XOXOXOXOXOXOXOXOXOXOXOXOXOXOXOXO")
    print("O                              X")
    print("X     Noughts and Crosses      O")
    print("O                              X")
    print("XOXOXOXOXOXOXOXOXOXOXOXOXOXOXOXO")
    print("")
    

def displayGrid(grid):
    print(" " + grid[0][0] + " | " + grid[0][1] + " | " + grid[0][2])
    print("-----------")
    print(" " + grid[1][0] + " | " + grid[1][1] + " | " + grid[1][2])
    print("-----------")
    print(" " + grid[2][0] + " | " + grid[2][1] + " | " + grid[2][2])    

def checkGrid(grid):
    #Checking the first row...
    if grid[0][0]=="X" and grid[0][1]=="X" and grid[0][2]=="X":
        print("Three Xs in a row.")
        game_over = 'won'
    elif grid[0][0]=="O" and grid[0][1]=="O" and grid[0][2]=="O":
        print("Three Os in a row.")
        game_over = 'lost'
    #Checking the second row...
    elif grid[1][0]=="X" and grid[1][1]=="X" and grid[1][2]=="X":
        print("Three Xs in a row.")
        game_over = 'won'
    elif grid[1][0]=="O" and grid[1][1]=="O" and grid[1][2]=="O":
        print("Three Os in a row.")
        game_over = 'lost'
    #Checking the third row...
    elif grid[2][0]=="X" and grid[2][1]=="X" and grid[2][2]=="X":
        print("Three Xs in a row.")
        game_over = 'won'
    elif grid[2][0]=="O" and grid[2][1]=="O" and grid[2][2]=="O":
        print("Three Os in a row.")
        game_over = 'lost'

    #Checking the first column
    elif grid[0][0]=="X" and grid[1][0]=="X" and grid[2][0]=="X":
        print("Three Xs in a column.")
        game_over = 'won'
    elif grid[0][0]=="O" and grid[1][0]=="O" and grid[2][0]=="O":
        print("Three Os in a column.")
        game_over = 'lost'
    #Checking the second column
    elif grid[0][1]=="X" and grid[1][1]=="X" and grid[2][1]=="X":
        print("Three Xs in a column.")
        game_over = 'won'
    elif grid[0][1]=="O" and grid[1][1]=="O" and grid[2][1]=="O":
        print("Three Os in a column.")
        game_over = 'lost'
    #Checking the third column
    elif grid[0][2]=="X" and grid[1][2]=="X" and grid[2][2]=="X":
        print("Three Xs in a column.")
        game_over = 'won'
    elif grid[0][2]=="O" and grid[1][2]=="O" and grid[2][2]=="O":
        print("Three Os in a column.")
        game_over = 'lost'
        
    #Checking the first diagonale
    elif grid[0][0]=="X" and grid[1][1]=="X" and grid[2][2]=="X":
        print("Three Xs in a diagonal.")
        game_over = 'won'
    elif grid[0][0]=="O" and grid[1][1]=="O" and grid[2][2]=="O":
        print("Three Os in a diagonal.")
        game_over = 'lost'
    #Checking the second diagonale
    elif grid[2][0]=="X" and grid[1][1]=="X" and grid[0][2]=="X":
        print("Three Xs in a diagonal.")
        game_over = 'won'
    elif grid[2][0]=="O" and grid[1][1]=="O" and grid[0][2]=="O":
        print("Three Os in a diagonal.")
        game_over = 'lost'
    else:
        game_over = False
        
    return game_over

def check_winner(game_over, free_cells):
    if game_over=='won':
        print('YOU WON!')
        return True
    elif game_over=='lost':
        print('YOU LOST!')
        return True
    elif len(free_cells)==0:
        print('YOU DRAW!')
        return True
    else:
        return False
```

```python
import time
import random
from IPython.display import clear_output

def tris():
    displayBanner()

    grid = [[" "," "," "],
            [" "," "," "],
            [" "," "," "]]

    displayGrid(grid)

    x_coord, y_coord = '', ''
    game_over = False
    counter = 0
    free_cells = [(x_coord, y_coord) for x_coord in range(3) for y_coord in range(3) if grid[x_coord][y_coord] == ' ']


    while game_over !='won' and game_over !='lost':
        if counter%2 ==0: # user turn
            while x_coord!=str(0) and x_coord!=str(1) and x_coord!=str(2):
                x_coord = input('Insert the x coordinate of your move (0-1-2)')
            while y_coord!=str(0) and y_coord!=str(1) and y_coord!=str(2):
                y_coord = input('Insert the y coordinate of your move (0-1-2)')

            if (int(x_coord),int(y_coord)) in free_cells:
                grid[int(x_coord)][int(y_coord)] = 'X'
                free_cells.remove((int(x_coord),int(y_coord)))
                counter+=1
                clear_output()
                displayBanner()
                displayGrid(grid)
                game_over = checkGrid(grid)
                if check_winner(game_over, free_cells):                
                    break
            else:
                print('Select a free cell')
                x_coord, y_coord = '', ''
        else: # computer turn
            print('Computer turn')
            time.sleep(2)
            comp_choice = random.choice(free_cells)
            x_coord = comp_choice[0]
            y_coord = comp_choice[1]
            grid[x_coord][y_coord] = 'O'
            free_cells.remove((x_coord,y_coord))
            x_coord, y_coord = '', ''
            counter+=1
            clear_output()
            displayBanner()
            displayGrid(grid)
            game_over = checkGrid(grid)
            if check_winner(game_over, free_cells):  
                break

```

```python
tris()
```
